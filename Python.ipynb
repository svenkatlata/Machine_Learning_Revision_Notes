## Casting

**Casting in Python** refers to converting a variable from one data type to another. Python supports both **explicit casting** (using built-in functions) and **implicit casting** (automatic type conversion by Python during certain operations).

---

### **1. Explicit Casting (Type Conversion)**

Python provides several built-in functions to convert data types explicitly.

#### **a. To Integer (`int()`)**
- Converts a value to an integer. Non-numeric strings and invalid inputs will raise a `ValueError`.

**Examples:**
```python
# From float to int
print(int(5.8))       # 5

# From string to int
print(int("42"))      # 42

# Invalid conversion
# print(int("hello")) # ValueError
```

---

#### **b. To Float (`float()`)**
- Converts a value to a floating-point number.

**Examples:**
```python
# From int to float
print(float(3))       # 3.0

# From string to float
print(float("5.67"))  # 5.67
```

---

#### **c. To String (`str()`)**
- Converts a value to a string.

**Examples:**
```python
# From int/float to string
print(str(42))        # '42'
print(str(3.14))      # '3.14'

# From other types to string
print(str(True))      # 'True'
```

---

#### **d. To Boolean (`bool()`)**
- Converts a value to a boolean (`True` or `False`).
- Non-zero numbers, non-empty strings, and non-empty collections are `True`; others are `False`.

**Examples:**
```python
# From numbers
print(bool(1))        # True
print(bool(0))        # False

# From strings
print(bool("hello"))  # True
print(bool(""))       # False
```

---

#### **e. To List (`list()`)**
- Converts a value to a list. Strings are split into individual characters, and other iterables are converted to lists.

**Examples:**
```python
# From string to list
print(list("hello"))  # ['h', 'e', 'l', 'l', 'o']

# From tuple to list
print(list((1, 2, 3)))  # [1, 2, 3]
```

---

#### **f. To Tuple (`tuple()`)**
- Converts a value to a tuple.

**Examples:**
```python
# From list to tuple
print(tuple([1, 2, 3]))  # (1, 2, 3)
```

---

#### **g. To Set (`set()`)**
- Converts a value to a set (removing duplicates if applicable).

**Examples:**
```python
# From list to set
print(set([1, 2, 2, 3]))  # {1, 2, 3}
```

---

#### **h. To Dictionary (`dict()`)**
- Converts a sequence of key-value pairs into a dictionary.

**Examples:**
```python
# From list of tuples to dict
print(dict([("a", 1), ("b", 2)]))  # {'a': 1, 'b': 2}
```

---

### **2. Implicit Casting (Type Conversion)**

Python automatically converts one data type to another where necessary and safe, typically during mathematical or comparison operations.

**Examples:**
```python
# From int to float (implicit)
x = 5
y = 2.5
result = x + y
print(result)        # 7.5 (float)

# From bool to int (implicit)
x = True + 3
print(x)             # 4
```

---

### **3. Common Use Cases**

#### **Mathematical Operations**
- Combining integers and floats automatically results in a float.
```python
print(5 + 3.0)       # 8.0
```

#### **String Concatenation**
- Explicit casting is needed to concatenate non-string types with strings.
```python
x = 10
print("Value: " + str(x))  # Value: 10
```

#### **User Input**
- Inputs from `input()` are strings by default and often require casting.
```python
age = int(input("Enter your age: "))
print(f"You are {age} years old.")
```

---

### **4. Best Practices**

1. **Use Explicit Casting**: Always cast explicitly to avoid unexpected results.
   ```python
   result = int("42") + 10
   ```

2. **Validate Inputs**: Check data types before casting to prevent runtime errors.
   ```python
   user_input = "42"
   if user_input.isdigit():
       age = int(user_input)
   ```

3. **Be Cautious with Float to Int**: Converting floats to integers truncates the decimal part.
   ```python
   print(int(5.9))  # 5
   ```

4. **Use `try-except` for Error Handling**: Handle exceptions during invalid casting.
   ```python
   try:
       x = int("hello")
   except ValueError:
       print("Invalid input!")
   ```

### Formated Strings and Raw Strings

#### **1. Formatted Strings (f-strings)**

**Formatted strings** (or **f-strings**) are a concise and readable way to include expressions inside string literals.

##### **Key Features**
- Prefix the string with `f` or `F`.
- Use curly braces `{}` to include expressions or variables.
- Expressions inside `{}` are evaluated and replaced with their values.

##### **Syntax**
```python
f"text {expression}"
```

##### **Examples**
1. **Variable Substitution**
   ```python
   name = "Alice"
   age = 25
   print(f"My name is {name} and I am {age} years old.")
   # Output: My name is Alice and I am 25 years old.
   ```

2. **Expressions Inside `{}`**
   ```python
   width = 5
   height = 10
   print(f"Area is {width * height}.")
   # Output: Area is 50.
   ```

3. **Formatting Numbers**
   ```python
   pi = 3.14159
   print(f"Pi rounded to 2 decimals: {pi:.2f}")
   # Output: Pi rounded to 2 decimals: 3.14
   ```

4. **Using Dictionaries**
   ```python
   data = {'name': 'Bob', 'age': 30}
   print(f"{data['name']} is {data['age']} years old.")
   # Output: Bob is 30 years old.
   ```

5. **Multiline f-strings**
   ```python
   name = "Charlie"
   print(f"""
   Hello, {name}!
   Welcome to Python.
   """)
   # Output:
   # Hello, Charlie!
   # Welcome to Python.
   ```

##### **Advantages**
- Highly readable and expressive.
- No need for string concatenation or calling `.format()`.

#### **2. Raw Strings**

**Raw strings** are useful when working with strings containing backslashes (`\`) that should not be treated as escape sequences.

##### **Key Features**
- Prefix the string with `r` or `R`.
- Backslashes (`\`) are treated as literal characters.
- Escape sequences like `\n`, `\t`, `\u` are not processed.

##### **Syntax**
```python
r"text with \backslashes"
```

##### **Examples**
1. **Literal Backslashes**
   ```python
   path = r"C:\Users\Alice\Documents"
   print(path)
   # Output: C:\Users\Alice\Documents
   ```

2. **Regular Expressions**
   ```python
   import re
   pattern = r"\d{3}-\d{2}-\d{4}"  # Matches a Social Security Number
   match = re.match(pattern, "123-45-6789")
   print(match is not None)
   # Output: True
   ```

3. **Avoiding Escape Interpretation**
   ```python
   print(r"Hello\nWorld")
   # Output: Hello\nWorld
   ```

##### **Limitations**
- A raw string cannot end with a single backslash (`\`) because the backslash escapes the closing quote.
   ```python
   # This will cause a SyntaxError
   raw_str = r"This ends with a backslash\"
   ```

### Mutability and Immutability

**Immutable Objects**: These objects **cannot** be modified after they are created. If you try to change their value, a new object is created.

**Mutable Objects**: These objects **can** be modified after they are created. Their state can be changed without creating a new object.

| **Category**     | **Immutable Objects**             | **Mutable Objects**              |
|-------------------|------------------------------------|------------------------------------|
| Numbers          | `int`, `float`, `complex`         | N/A                                |
| Sequences        | `str`, `tuple`, `bytes`           | `list`, `bytearray`               |
| Sets             | `frozenset`                       | `set`                             |
| Mappings         | N/A                               | `dict`                            |
| Others           | Custom immutable types            | Custom mutable types              |


### String Indexing and Slicing

````python
# Sample string to use
sample_string = "Hello, World!"

# Basic slicing: [start:stop:step]

# Examples
example_1 = sample_string[0:5]  # Example 1: Slice from index 0 to 5 (exclusive)
print("Example 1:", example_1)  # 'Hello'

example_2 = sample_string[7:]  # Example 2: Slice from index 7 to the end
print("Example 2:", example_2)  # 'World!'

example_3 = sample_string[:5]  # Example 3: Slice from start to index 5 (exclusive)
print("Example 3:", example_3)  # 'Hello'

example_4 = sample_string[:]  # Example 4: Full slice (entire string)
print("Example 4:", example_4)  # 'Hello, World!'

example_5 = sample_string[::2]  # Example 5: Slice with a step of 2
print("Example 5:", example_5)  # 'Hlo ol!'

example_6 = sample_string[::-1]  # Example 6: Slice with a negative step (reverse string)
print("Example 6:", example_6)  # '!dlroW ,olleH'

example_7 = sample_string[1:10:2]  # Example 7: Slice with a specific range and step
print("Example 7:", example_7)  # 'el,Wr'

example_8 = sample_string[-6:-1]  # Example 8: Negative indices to slice
print("Example 8:", example_8)  # 'World'

example_9 = sample_string[-1:-6:-1]  # Example 9: Negative indices with step
print("Example 9:", example_9)  # '!dlro'

example_10 = sample_string[::3]  # Example 10: Omit start and stop with step
print("Example 10:", example_10)  # 'Hl r!'

example_11 = sample_string[5::-1]  # Example 11: Omit stop with negative step
print("Example 11:", example_11)  # 'o lleH'

example_12 = sample_string[-5:12]  # Example 12: Combine positive and negative indices
print("Example 12:", example_12)  # 'orld'

example_13 = sample_string[5:5]  # Example 13: Slice with no range
print("Example 13:", example_13)  # ''

example_14 = sample_string[0:5:10]  # Example 14: Step larger than slice range
print("Example 14:", example_14)  # 'H'

example_15 = sample_string[5:0:-1]  # Example 15: Negative step with partial range
print("Example 15:", example_15)  # 'o ll'

example_16 = sample_string[0:50]  # Example 16: Out-of-bounds slice
print("Example 16:", example_16)  # 'Hello, World!'

example_17 = sample_string[-50:-1]  # Example 17: Negative out-of-bounds slice
print("Example 17:", example_17)  # 'Hello, World'

empty_string = ""
example_18 = empty_string[:5]  # Example 18: Slicing an empty string
print("Example 18:", example_18)  # ''
````

### Assignment, Shallow Copy and Deep Copy of Lists

#### 1. **Assignment (`list2 = list1`)**
- **Description**: `list2` becomes just a reference to the same object as `list1`. Modifications to `list2` will also affect `list1`.
- **Example**:
  ```python
  list1 = [1, 2, [3, 4]]
  list2 = list1
  list2[0] = 99
  print(list1)  # Output: [99, 2, [3, 4]]
  ```

---

#### 2. **Shallow Copy (`list2 = list1.copy()` or `list2 = list(list1)` or `list2 = list1[:]`)**
- **Description**: Creates a new list object. The outer list is copied, but inner objects (nested lists, etc.) are still referenced. Changes to inner objects will reflect in both lists.
- **Example**:
  ```python
  import copy
  
  list1 = [1, 2, [3, 4]]
  list2 = list1.copy()  # Shallow copy
  list2[0] = 99
  list2[2][0] = 88
  print(list1)  # Output: [1, 2, [88, 4]] (Inner list modified)
  print(list2)  # Output: [99, 2, [88, 4]] (Outer list independent)
  ```

---

#### 3. **Deep Copy (`copy.deepcopy(list1)`)**
- **Description**: Creates a new list object, including recursively copying all nested objects. Changes to the copy will not affect the original list.
- **Example**:
  ```python
  import copy
  
  list1 = [1, 2, [3, 4]]
  list2 = copy.deepcopy(list1)  # Deep copy
  list2[0] = 99
  list2[2][0] = 88
  print(list1)  # Output: [1, 2, [3, 4]] (Unchanged)
  print(list2)  # Output: [99, 2, [88, 4]] (Independent)
  ```

---

<br>

| Method        | Outer List | Inner Objects (Nested Lists) |
|---------------|------------|------------------------------|
| Assignment (`list2 = list1`)  | Shared      | Shared                       |
| Shallow Copy (`list2 = list1[:]`) | Independent | Shared                       |
| Deep Copy (`list2 = copy.deepcopy(list1)`) | Independent | Independent                  |

Use **shallow copy** for flat lists and **deep copy** for nested lists when true independence is required.

### List Methods

````python
# Initial list for demonstration
my_list = [10, 20, 30, 40, 50]

# 1. append() - Add a single element at the end
my_list.append(60)
print("After append:", my_list)  # Output: [10, 20, 30, 40, 50, 60]

my_list.append([70, 80])
print("After append a list:", my_list)  # Output: [10, 20, 30, 40, 50, 60, [70, 80]]

del my_list[-1]
print("After deletion:", my_list)  # Output: [10, 20, 30, 40, 50, 60]

# 2. extend() - Add all elements of another iterable to the end
my_list.extend([70, 80])
print("After extend:", my_list)  # Output: [10, 20, 30, 40, 50, 60, 70, 80]

# 3. insert() - Insert an element at a specific index
my_list.insert(2, 25)
print("After insert:", my_list)  # Output: [10, 20, 25, 30, 40, 50, 60, 70, 80]

# 4. remove() - Remove the first occurrence of a specific value
my_list.remove(30)
print("After remove:", my_list)  # Output: [10, 20, 25, 40, 50, 60, 70, 80]

# 5. pop() - Remove and return an element by index (last by default)
removed = my_list.pop()
print("After pop:", my_list)  # Output: [10, 20, 25, 40, 50, 60, 70]
print("Popped element:", removed)  # Output: 80

# 6. index() - Return the index of the first occurrence of a value
index = my_list.index(40)
print("Index of 40:", index)  # Output: 3

# 7. count() - Return the count of occurrences of a value
count = my_list.count(20)
print("Count of 20:", count)  # Output: 1

# 8. sort() - Sort the list in ascending order
my_list.sort()
print("After sort:", my_list)  # Output: [10, 20, 25, 40, 50, 60, 70]

# 9. reverse() - Reverse the order of the list
my_list.reverse()
print("After reverse:", my_list)  # Output: [70, 60, 50, 40, 25, 20, 10]

# 10. copy() - Return a shallow copy of the list
copied_list = my_list.copy()
print("Copied list:", copied_list)  # Output: [70, 60, 50, 40, 25, 20, 10]

# 11. clear() - Remove all elements from the list
my_list.clear()
print("After clear:", my_list)  # Output: []
````

### List Unpacking

List unpacking in Python refers to assigning the elements of a list to individual variables. Below are examples showcasing various ways to use list unpacking:

---

#### **1. Basic Unpacking**
Assign elements of a list to variables.
```python
numbers = [1, 2, 3]
a, b, c = numbers
print(a, b, c)  # Output: 1 2 3
```

---

#### **2. Unpacking with `*` (Star Operator)**
Using `*` allows capturing multiple elements in a list.
```python
numbers = [1, 2, 3, 4, 5]

# Capture the first two elements, and the rest in another list
a, b, *rest = numbers
print(a, b, rest)  # Output: 1 2 [3, 4, 5]

# Capture all elements except the last
*rest, last = numbers
print(rest, last)  # Output: [1, 2, 3, 4] 5

# Capture the first, last, and everything in between
first, *middle, last = numbers
print(first, middle, last)  # Output: 1 [2, 3, 4] 5
```

---

#### **3. Unpacking with Arbitrary Length Lists**
Unpacking works with lists of varying lengths if you handle the elements appropriately.
```python
numbers = [10, 20, 30]

# Only unpack the first two elements
a, b, *_ = numbers
print(a, b)  # Output: 10 20
```

---

#### **4. Nested List Unpacking**
Unpack nested lists into variables.
```python
nested_list = [1, [2, 3], 4]

a, (b, c), d = nested_list
print(a, b, c, d)  # Output: 1 2 3 4
```

---

#### **5. Unpacking with Loops**
Iterate over lists with unpacking.
```python
pairs = [(1, 2), (3, 4), (5, 6)]

for x, y in pairs:
    print(x, y)
# Output:
# 1 2
# 3 4
# 5 6
```

---

#### **6. Unpacking Strings from a List**
Lists containing strings can also be unpacked.
```python
words = ["hello", "world"]

a, b = words
print(a, b)  # Output: hello world
```

---

#### **7. Ignoring Elements While Unpacking**
Use `_` to ignore elements you dont need.
```python
numbers = [1, 2, 3, 4]

a, _, c, _ = numbers
print(a, c)  # Output: 1 3
```

---

#### **8. Using List Unpacking in Function Arguments**
You can pass unpacked lists as arguments to functions.
```python
def add(x, y, z):
    return x + y + z

numbers = [1, 2, 3]
result = add(*numbers)  # Unpack the list into individual arguments
print(result)  # Output: 6
```

---

#### **9. Merging Lists with Unpacking**
Unpack lists to merge them.
```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

merged = [*list1, *list2]
print(merged)  # Output: [1, 2, 3, 4, 5, 6]
```

---

#### **10. Working with Mixed Data Types**
Unpack lists with mixed data types.
```python
data = [1, "hello", [3, 4], (5, 6)]

a, b, c, d = data
print(a, b, c, d)  # Output: 1 hello [3, 4] (5, 6)
```

### Dictionary Unpacking

#### **1. Unpacking for Function Arguments**
You can pass dictionary keys and values as arguments to a function using the `**` operator.

```python
def greet(name, age):
```

#### **2. Merging Dictionaries**
Unpacking can merge multiple dictionaries into one (available in Python 3.5+).

##### **Example 1: Using ** Operator**
```python
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

# Merge dictionaries
merged = {**dict1, **dict2}
print(merged)
# Output: {'a': 1, 'b': 3, 'c': 4}
```

The value of key `b` in `dict2` overwrites the value from `dict1`.

##### **Example 2: Using `dict.update()`**
```python
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

dict1.update(dict2)
print(dict1)
# Output: {'a': 1, 'b': 3, 'c': 4}
```

#### **3. Creating a New Dictionary**
You can use unpacking to combine dictionaries while adding or modifying key-value pairs.

```python
dict1 = {"a": 1, "b": 2}

# Output: {'a': 1, 'b': 4, 'c': 3}
```

```python
data = {"b": 2, "a": 1, "c": 3}
```

### **`Match-Case` Conditions**

`match-case` is a feature introduced in Python 3.10 that provides a concise and readable way to perform pattern matching, similar to `switch-case` in other programming languages but with added power. It can be used for checking conditions and matching complex patterns in a structured and intuitive manner.

#### 1. Basic Matching
```python
status = "pending"

match status:
    case "approved":
        print("The status is approved")
    case "rejected":
        print("The status is rejected")
    case "pending":
        print("The status is pending")
    case _:    # Wildcards - The underscore `_` can be used as a wildcard to ignore parts of the structure.
        print("Unknown status")
```

#### 2. Matching with Conditions
You can add conditions (also called guards) to a `case` statement:
```python
number = 42

match number:
    case n if n < 0:
        print("Negative number")
    case n if n == 0:
        print("Zero")
    case n if n > 0:
        print("Positive number")
    case _:    
        print("Unknown")
```

#### 3. Matching Multiple Values
You can match multiple possible values in one case:
```python
value = "yes"

match value:
    case "yes" | "y" | "Y":
        print("Affirmative response")
    case "no" | "n" | "N":
        print("Negative response")
    case _:
        print("Unrecognized response")
```

#### 4. Structured Patterns (Decomposing Data)
`match-case` can be used with structured data like tuples, lists, or dictionaries.
```python
point = (0, 10)

match point:
    case (0, y):
        print(f"Point is on the Y-axis at y = {y}")
    case (x, 0):
        print(f"Point is on the X-axis at x = {x}")
    case (x, y):
        print(f"Point is at ({x}, {y})")
```

#### 5. Matching Dictionaries
With `match-case`, you can even match dictionary-like patterns.
```python
data = {"type": "error", "message": "Something went wrong"}

match data:
    case {"type": "error", "message": msg}:
        print(f"Error: {msg}")
    case {"type": "success", "data": result}:
        print(f"Success: {result}")
    case _:
        print("Unknown response")
```

#### **Short-Circuit Evaluation**
- **`and`**: Stops evaluating as soon as it encounters a `False` value because the entire expression cannot be `True`.
- **`or`**: Stops evaluating as soon as it encounters a `True` value because the entire expression cannot be `False`.

```python
# Short-circuit with or

```

### is vs ==

````python
# == (Equality Operator): Compares values of objects, checks if they are equal in terms of content.
print("Equality Operator (==):")
x = [1, 2, 3]
y = [1, 2, 3]
print(x == y)  # Output: True, because both lists have the same values

z = [1, 2, 3]
print(x == z)  # Output: True, because both lists have the same values

# is (Identity Operator): Compares the identity of objects, checks if they are the same object in memory.
print("\nIdentity Operator (is):")
x = [1, 2, 3]
y = [1, 2, 3]
print(x is y)  # Output: False, because x and y refer to different objects in memory

z = x
print(x is z)  # Output: True, because both x and z refer to the same object
````

### **Arguments vs Parameters**

The terms **arguments** and **parameters** are often used interchangeably, but they have distinct meanings in the context of programming, especially in Python.

---

#### **Definition**

1. **Parameter**:
   - A variable defined in the function signature.
   - Acts as a placeholder for the value that will be passed to the function.
   - Exists only during the function's execution.

   Example:
   ```python
   def greet(name):  # 'name' is the parameter
       print(f"Hello, {name}!")
   ```

2. **Argument**:
   - The actual value passed to the function when it is called.
   - It replaces the parameter during function execution.

   Example:
   ```python
   greet("Alice")  # "Alice" is the argument
   ```

---

#### **Key Difference**

| **Aspect**      | **Parameter**                        | **Argument**                   |
|------------------|--------------------------------------|---------------------------------|
| **Definition**   | Variable in the function definition. | Value passed during the function call. |
| **When Used**    | During function definition.          | During function invocation.    |
| **Example**      | `name` in `def greet(name):`.        | `"Alice"` in `greet("Alice")`. |

---


#### **Types of Arguments**

1. **Positional Arguments**:
   - Matched to parameters based on their position.
   
   ```python
   def greet(first_name, last_name):
       print(f"Hello, {first_name} {last_name}!")
       
   greet("Alice", "Smith")  # Positional arguments
   ```

2. **Keyword Arguments**:
   - Matched to parameters using the parameter name.

   ```python
   greet(first_name="Alice", last_name="Smith")  # Keyword arguments
   ```

3. **Default Arguments**:
   - Provide default values for parameters.

   ```python
   def greet(name="Guest"):
       print(f"Hello, {name}!")

   greet()  # Uses the default argument "Guest"
   greet("Alice")  # Overrides the default
   ```

4. **Variable-Length Arguments**:
   - Pass a varying number of arguments using `*args` or `**kwargs`.

   ```python
   def print_numbers(*args):  # 'args' is a tuple of positional arguments
       for num in args:
           print(num)

   print_numbers(1, 2, 3, 4)
   ```

   ```python
   def print_info(**kwargs):  # 'kwargs' is a dictionary of keyword arguments
       for key, value in kwargs.items():
           print(f"{key}: {value}")

   print_info(name="Alice", age=30)
   ```

```python
print(greet("Alice"))  # Output: Hello, Alice!
```

- **Key Characteristics:**

  - Can be used without associating it with a specific object.

### Walrus Operator (:=)

The **walrus operator** (`:=`), introduced in Python 3.8, is an **assignment expression** that allows you to assign a value to a variable as part of an expression. It combines assignment and evaluation into a single step.

---

#### **Syntax**
```python
variable := expression
```

- **`variable`**: The name of the variable being assigned.
- **`expression`**: The value or expression being evaluated and assigned to the variable.

---

#### **Key Benefits**
1. Reduces redundancy by allowing assignment inside expressions.
2. Improves readability and performance by avoiding repeated computations.

---

#### **Examples**

##### **1. Simplifying While Loops**
Without the walrus operator:
```python
line = input("Enter text: ")
while line != "quit":
    print(f"You entered: {line}")
    line = input("Enter text: ")
```

With the walrus operator:
```python
while (line := input("Enter text: ")) != "quit":
    print(f"You entered: {line}")
```

---

##### **2. Reducing Redundant Computations**
Without the walrus operator:
```python
nums = [1, 2, 3, 4, 5]
squared = [x**2 for x in nums if x**2 > 10]
print(squared)
```

With the walrus operator:
```python
nums = [1, 2, 3, 4, 5]
squared = [square for x in nums if (square := x**2) > 10]
print(squared)
```

---

##### **3. Combining Assignment with Conditional Statements**
Without the walrus operator:
```python
value = len(my_list)
if value > 5:
    print(f"The list is too long: {value}")
```

With the walrus operator:
```python
if (value := len(my_list)) > 5:
    print(f"The list is too long: {value}")
```

---

#### **Use Cases**
1. **While loops**: Avoid pre-loop initialization and redundant re-evaluations.
2. **List comprehensions**: Store intermediate results for filtering or transformations.
3. **Conditional statements**: Simplify expressions that involve a computed value.

---

#### **Things to Note**
1. The walrus operator **does not replace the normal assignment (`=`)**. It is used only in expressions, not standalone assignments.
   ```python
   x := 5  #  SyntaxError
   x = 5   #  Correct
   ```
2. **Improper use may reduce readability.** Use it sparingly in complex expressions to keep your code understandable.

### **Scope in Python**

**Scope** refers to the region in a program where a variable or object is accessible. In Python, the scope determines which parts of the code can access a particular variable.

---

#### **Types of Scope in Python**

1. **Local Scope**  
   Variables declared inside a function or block are accessible only within that function or block.  
   These variables are **local variables**.

   ```python
   def greet():
       message = "Hello!"  # Local variable
       print(message)

   greet()
   # print(message)  #  NameError: message is not defined
   ```

2. **Global Scope**  
   Variables declared outside of all functions or blocks are accessible throughout the program.  
   These are **global variables**.

   ```python
   global_var = "I am global!"  # Global variable

   def display():
       print(global_var)  # Accessible inside the function

   display()
   print(global_var)  # Accessible outside the function
   ```

3. **Enclosed Scope (Nonlocal)**  
   Variables in the enclosing (outer) function of a nested function are in an **enclosed scope**.  
   They are neither local nor global to the nested function.

   ```python
   def outer_function():
       outer_var = "Outer variable"

       def inner_function():
           print(outer_var)  # Accessing enclosed variable
       
       inner_function()

   outer_function()
   ```

4. **Built-in Scope**  
   Built-in functions, exceptions, and objects provided by Python (like `len`, `print`, etc.) belong to the **built-in scope** and are available everywhere.

   ```python
   print(len([1, 2, 3]))  # Built-in function
   ```

---

### **LEGB Rule**  

Python resolves variable names using the **LEGB** rule:

1. **Local**: Look in the current function's local scope.  
2. **Enclosed**: Look in the enclosing scope (e.g., outer functions).  
3. **Global**: Look in the global scope.  
4. **Built-in**: Look in the built-in scope.

---

#### **Examples**

##### **LEGB in Action**
```python
x = "global"

def outer():
    x = "enclosed"
    def inner():
        x = "local"
        print(x)  # Prints "local"
    inner()

outer()
print(x)  # Prints "global"
```

---

#### **Global Keyword**

The `global` keyword allows you to modify a global variable from within a function.

```python
x = 10  # Global variable

def modify_global():
    global x
    x = 20  # Modifying the global variable

modify_global()
print(x)  # Output: 20
```

---

#### **Nonlocal Keyword**

The `nonlocal` keyword is used to modify a variable in the enclosing (non-global) scope.

```python
def outer():
    x = "outer"
    def inner():
        nonlocal x
        x = "modified"
    inner()
    print(x)  # Output: "modified"

outer()
```

---

#### **Key Points**

1. Local variables have the narrowest scope and are temporary.
2. Global variables are accessible across the entire script but can be modified using the `global` keyword.
3. Enclosed variables exist in nested functions and can be modified with the `nonlocal` keyword.
4. Python follows the **LEGB rule** for resolving variable names.

---

### **Q. Why Do We Need Scope in Programming?**

Scope is essential in programming because it controls the visibility and lifetime of variables. Without scope, managing variables, understanding code, and avoiding errors in large programs would become significantly more challenging. Here's why scope is important:

---

#### **1. Prevents Variable Name Conflicts**
- Scope ensures that variables in one part of a program do not accidentally overwrite or interfere with variables in another part.
- For example, local variables in functions can have the same name without conflicting with global variables.

```python
x = 10  # Global variable

def my_function():
    x = 5  # Local variable
    print(x)  # Refers to the local x

my_function()  # Output: 5
print(x)        # Output: 10 (global x is unchanged)
```

---

#### **2. Improves Code Readability and Organization**
- Scope helps you organize code into meaningful, self-contained blocks (e.g., functions or classes).
- Variables declared within a specific scope are used only where needed, making the code easier to understand.

---

#### **3. Enables Reusability of Functions and Code Blocks**
- By defining variables locally, you can write reusable functions and modules that don't depend on global state.
- This makes your code more modular and easier to debug.

```python
def add_numbers(a, b):
    result = a + b  # Local variables
    return result

print(add_numbers(3, 4))  # Output: 7
```

---

#### **4. Enhances Security and Encapsulation**
- Variables with restricted scope (e.g., local or enclosed variables) cannot be accessed or modified outside their defined area.
- This reduces unintended interactions and protects sensitive data.

```python
def secure_function():
    secret = "This is hidden"  # Local variable
    return "Access granted"

# print(secret)  #  NameError: 'secret' is not defined outside the function
```

---

#### **5. Optimizes Memory Usage**
- Variables in a narrower scope (e.g., local variables) exist only during the execution of their block and are automatically discarded afterward.
- This efficient memory management prevents the accumulation of unused data.

```python
def temporary_scope():
    temp_var = "I exist temporarily"
    print(temp_var)

temporary_scope()
# temp_var no longer exists here
```

---

#### **6. Facilitates Debugging**
- Scoping restricts the availability of variables, so when debugging, you only need to focus on the relevant variables within a specific scope.
- This makes identifying and fixing issues easier.

---

#### **7. Enables Context-Specific Behavior**
- Scopes allow the same variable name to have different meanings in different contexts (e.g., nested functions or loops).

```python
def outer():
    x = "outer variable"
    def inner():
        x = "inner variable"
        print(x)  # Refers to the inner x
    inner()
    print(x)  # Refers to the outer x

outer()
```

---

#### **8. Supports the LEGB Rule**
- Python's **LEGB rule** (Local, Enclosed, Global, Built-in) enables structured name resolution by searching for variables in a logical order.
- This ensures that Python code runs predictably.

---

#### **Without Scope, What Happens?**
- Variables could be accessible everywhere, leading to:
  - **Unexpected modifications**: Variables may be unintentionally changed in unrelated parts of the program.
  - **Poor modularity**: Functions and blocks would rely on external variables, reducing reusability.
  - **Memory leaks**: Variables would persist longer than needed, wasting memory.

---

#### **In Summary**
Scope is a fundamental concept that:
- Ensures clean and organized code.
- Prevents conflicts and protects variables.
- Optimizes memory and debugging efforts.
- Allows you to write secure, reusable, and modular programs.

# OOP

Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects," which can contain data (attributes or properties) and code (methods or functions). It is designed to help developers organize and structure code in a more intuitive and reusable way. Here are the 4 Pillars of OOP:

1. **Encapsulation**: The concept of bundling data (attributes) and methods (functions) into a single unit called a class. It also involves restricting direct access to some of the object's data, which can be done by making attributes private and providing public methods (getters and setters) to access them.
   - Example: A class may have private variables, and only methods are used to get or set their values.

2. **Abstraction**: Hiding complex implementation details and providing a simplified interface for interacting with an object. This is often achieved by using abstract classes and interfaces.
   - Example: An abstract class `Shape` might define a method `draw()`, but the actual drawing behavior is defined in subclasses like `Circle` or `Rectangle`.

3. **Inheritance**: A mechanism where one class (called the child or subclass) inherits attributes and methods from another class (called the parent or superclass). This promotes code reuse and allows for the creation of hierarchies.
   - Example: A `Dog` class might inherit from an `Animal` class, meaning the `Dog` class will have properties and behaviors of the `Animal` class, like `eat()` and `sleep()`.

4. **Polymorphism**: The ability of different classes to be treated as instances of the same class through inheritance. It also allows methods to have the same name but behave differently based on the object they are acting upon (method overriding).
   - Example: A method `makeSound()` could be implemented differently in both the `Dog` and `Cat` classes, even though both classes use the same method name.

OOP is implemented using the following concepts:

1. **Classes**: A blueprint or template for creating objects. It defines the attributes and methods that an object will have.
   - Example: A `Car` class might define properties like `color` and `make`, and methods like `drive()` and `stop()`.

2. **Objects**: Instances of classes. They represent real-world entities and can have specific values for attributes.
   - Example: An object of the `Car` class might have the `color` attribute set to "red" and the `make` attribute set to "Toyota."

In OOP, classes and objects allow for the creation of more organized, maintainable, and scalable code. It emphasizes the reusability of code through inheritance and polymorphism, and encourages better organization of data and methods.

### Classes

A class is a blueprint for creating objects (instances). It defines the properties and behaviors of the objects.

- Naming Convention for all other variables, function names etc.: snake_case

### Attributes

In Python, **attributes** in classes refer to the variables or properties that are associated with an instance of a class (also known as **instance attributes**) or with the class itself (known as **class attributes**). These attributes hold data that describes the state of an object, and they can be accessed or modified using the object or class.

#### **Types of Attributes in Classes**
There are mainly two types of attributes in Python classes:

1. **Instance Attributes**
2. **Class Attributes**

---

#### 1. **Instance Attributes**
Instance attributes are variables that belong to a specific instance of the class. They are defined using the `self` keyword inside methods (usually the `__init__` constructor). Each object (instance) of the class will have its own copy of these attributes.

##### Example:
```python
class Dog:
    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age

# Creating instances of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Charlie", 5)

# Accessing instance attributes
print(dog1.name)  # Output: Buddy
print(dog2.age)   # Output: 5
```

In this example:
- `name` and `age` are instance attributes.
- Each instance (`dog1` and `dog2`) has its own values for `name` and `age`.

---

#### 2. **Class Attributes**
Class attributes are variables that belong to the class itself and are shared by all instances of the class. Class attributes are defined directly inside the class, outside of any methods. These attributes have the same value for all instances of the class, unless they are modified.

##### Example:
```python
class Dog:
    # Class attribute
    species = "Canis familiaris"

    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age

# Creating instances of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Charlie", 5)

# Accessing class attributes
print(Dog.species)  # Output: Canis familiaris
print(dog1.species)  # Output: Canis familiaris
print(dog2.species)  # Output: Canis familiaris
```

In this example:
- `species` is a class attribute.
- All instances (`dog1`, `dog2`, etc.) share the same value for `species`.

---

#### **Accessing and Modifying Attributes**
- **Accessing Instance Attributes:** You access instance attributes through the object (instance).
- **Accessing Class Attributes:** You can access class attributes through either the class name or the object. However, if you modify a class attribute using an instance, it will create a new instance attribute (it does not modify the class attribute).

##### Example of Accessing and Modifying Attributes:
```python
class Dog:
    species = "Canis familiaris"  # Class attribute

    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute

# Creating an instance
dog1 = Dog("Buddy", 3)

# Accessing attributes
print(dog1.species)  # Output: Canis familiaris (Access via instance)
print(Dog.species)   # Output: Canis familiaris (Access via class)

# Modifying instance attribute
dog1.age = 4
print(dog1.age)  # Output: 4

# Modifying class attribute
Dog.species = "Canis lupus"
print(dog1.species)  # Output: Canis lupus
print(Dog.species)   # Output: Canis lupus

# Modifying instance attributes without affecting class attribute
dog1.species = "Doggo"
print(dog1.species)  # Output: Doggo (Modified instance attribute)
print(Dog.species)   # Output: Canis lupus (Class attribute is unchanged)
```

In this example:
- Changing `dog1.species` modifies only the instance attribute `species`, not the class attribute.
- If you modify `Dog.species`, it affects all instances of the class (unless they have an instance-specific attribute).

### Methods

In Python, **methods** are functions that are defined inside a class and are used to define the behaviors of objects (instances) of that class. Methods are closely tied to the attributes of the class and allow objects to perform actions or manipulate their data.

#### **Types of Methods in Python Classes**

There are three main types of methods in Python classes:
1. **Instance Methods**
2. **Class Methods**
3. **Static Methods**

---

#### 1. **Instance Methods**
Instance methods are the most common methods and are defined within the class using the `def` keyword. They always take at least one parameter, `self`, which represents the instance of the class. This allows the method to access and modify the instance attributes.

##### Example:
```python
class Dog:
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute
    
    def bark(self):  # Instance method
        print(f"{self.name} says Woof!")

    def birthday(self):  # Instance method
        self.age += 1
        print(f"{self.name} is now {self.age} years old!")

# Creating an instance of Dog
dog1 = Dog("Buddy", 3)

# Calling instance methods
dog1.bark()         # Output: Buddy says Woof!
dog1.birthday()     # Output: Buddy is now 4 years old!
```

In this example:
- `bark` and `birthday` are instance methods that operate on instance attributes (`name` and `age`).
- They access the instance data via the `self` parameter.

---

#### 2. **Class Methods**
Class methods are defined using the `@classmethod` decorator. They take `cls` as their first parameter, which refers to the class itself (**not the instance**). Class methods are used for operations that affect the class as a whole, rather than individual instances.

#### Example:
```python
class Dog:
    species = "Canis familiaris"  # Class attribute

    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def change_species(cls, new_species):  # Class method
        cls.species = new_species
        print(f"Species changed to {cls.species}!")

# Creating an instance of Dog
dog1 = Dog("Buddy", 3)

# Calling the class method
Dog.change_species("Canis lupus")  # Output: Species changed to Canis lupus!
print(dog1.species)  # Output: Canis lupus (Modified for all instances)
```

In this example:
- `change_species` is a class method that modifies the class attribute `species`.
- Class methods can be called using the class name (`Dog.change_species`) or an instance (`dog1.change_species`).

---

### 3. **Static Methods**
Static methods are defined using the `@staticmethod` decorator. Unlike instance methods and class methods, static methods do not take `self` or `cls` as their first parameter. They behave like regular functions but are included within the class for organizational purposes. Static methods **do not access or modify instance or class attributes**.

#### Example:
```python
class Dog:
    @staticmethod
    def bark():  # Static method
        print("Woof!")

# Calling the static method
Dog.bark()  # Output: Woof!
```

In this example:
- `bark` is a static method. It is not tied to an instance or a class attribute.
- Static methods are typically used when the behavior does not depend on instance or class-specific data but still logically belongs within the class.

---

### Key Differences Between Instance Methods, Class Methods, and Static Methods:

| Feature            | Instance Method           | Class Method               | Static Method               |
|--------------------|---------------------------|----------------------------|-----------------------------|
| **Decorator**       | None                      | `@classmethod`             | `@staticmethod`             |
| **First parameter** | `self` (instance of class)| `cls` (class itself)        | None (no access to instance or class)|
| **Access**          | Can access instance attributes and methods | Can access class attributes and methods | Cannot access instance or class attributes |
| **Use case**        | Operates on instance data| Operates on class data      | Does not operate on instance or class data |

## Abstraction

### Access Modifiers

In Python, **access modifiers** (private, protected, and public) control the visibility and accessibility of class attributes (variables) and methods. Python uses a convention rather than strict enforcement for these access control levels. However, it provides a way to indicate whether a variable or method is intended to be private, protected, or public.

#### 1. **Public Variables**:
Public variables can be accessed directly from outside the class. By default, all attributes and methods are public in Python.

##### Example of Public Variables:
```python
class Car:
    def __init__(self, make, model):
        self.make = make  # Public attribute
        self.model = model  # Public attribute

# Creating an object of Car
my_car = Car("Toyota", "Corolla")

# Accessing public variables
print(my_car.make)  # Output: Toyota
print(my_car.model)  # Output: Corolla
```

- In this example, `make` and `model` are public variables, which can be accessed directly from outside the class.

#### 2. **Protected Variables**:
Protected variables are intended to be used **within the class and its subclasses**. In Python, there's no strict enforcement of access control, but we use a single underscore (`_`) as a convention to indicate that the variable is protected.

##### Example of Protected Variables:
```python
class Car:
    def __init__(self, make, model):
        self._make = make  # Protected attribute
        self._model = model  # Protected attribute

    def display_info(self):
        print(f"Car Make: {self._make}, Model: {self._model}")

# Creating an object of Car
my_car = Car("Toyota", "Corolla")

# Accessing protected variables directly (not recommended)
print(my_car._make)  # Output: Toyota
print(my_car._model)  # Output: Corolla

# It is advised not to access directly, but instead use a method
my_car.display_info()  # Output: Car Make: Toyota, Model: Corolla
```

- In this example, `_make` and `_model` are **protected variables**. These can be accessed from outside the class but are intended to be accessed by the class and its subclasses.
- **Note**: It is a convention that suggests "don't access directly," but Python doesn't prevent it.

#### 3. **Private Variables**:
Private variables are intended to be used **only within the class**. In Python, private variables are indicated by using a **double underscore (`__`)** before the variable name. This triggers name mangling, which changes the variable's name to prevent it from being easily accessed outside the class.

##### Example of Private Variables:
```python
class Car:
    def __init__(self, make, model):
        self.__make = make  # Private attribute
        self.__model = model  # Private attribute

    def display_info(self):
        print(f"Car Make: {self.__make}, Model: {self.__model}")

# Creating an object of Car
my_car = Car("Toyota", "Corolla")

# Accessing private variables directly will raise an AttributeError
# print(my_car.__make)  # This will raise an error

# Accessing private variables through a public method
my_car.display_info()  # Output: Car Make: Toyota, Model: Corolla
```

- In this example, `__make` and `__model` are private variables. They are not directly accessible from outside the class.
- **Name Mangling**: If you try to access a private variable directly like `my_car.__make`, Python will raise an `AttributeError`. However, Python internally "mangles" the name, changing it to `_Car__make` and `_Car__model`. So, it's still possible to access it through its mangled name, but it is not recommended.
  
  You can access the private variables using the mangled name (not recommended):
  ```python
  print(my_car._Car__make)  # Output: Toyota
  ```

---

#### Summary of Access Levels:
| Access Level    | Convention       | Access from Outside the Class | Access from Subclasses |
|-----------------|------------------|--------------------------------|------------------------|
| **Public**      | No underscore (`make`) | Can be accessed directly       | Can be accessed directly |
| **Protected**   | Single underscore (`_make`) | Can be accessed (by convention, should not) | Can be accessed          |
| **Private**     | Double underscore (`__make`) | Cannot be accessed directly   | Cannot be accessed directly (unless using name mangling) |

---

#### Best Practices:
- **Public variables** should be used for data that needs to be accessible by all parts of the program.
- **Protected variables** should be used for internal data that should not be directly accessed outside the class, but may be accessed by subclasses.
- **Private variables** should be used for data that should be completely hidden from outside access, including from subclasses.

While Python's use of underscores (`_`, `__`) is just a convention, respecting these conventions helps ensure cleaner, more maintainable code and prevents unintentional misuse of class attributes.

## Inheritance

**Inheritance** is one of the core concepts in Object-Oriented Programming (OOP). It allows a class (known as the **child class** or **subclass**) to inherit attributes and methods from another class (known as the **parent class** or **superclass**). This promotes **code reuse** and helps in building hierarchical relationships between classes.

#### Key Concepts:
1. **Parent Class (Superclass)**: The class from which other classes inherit.
2. **Child Class (Subclass)**: The class that inherits from the parent class and can override or extend its functionality.
3. **Method Overriding**: A child class can provide a specific implementation of a method that is already defined in the parent class.
4. **`super()` Function**: It allows a child class to call methods from the parent class.

````python
# Parent class (Superclass)
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound.")

# Child class (Subclass) that inherits from Animal
class Dog(Animal):
    def __init__(self, name, breed):
        # Calling the parent class's constructor (optional)
        super().__init__(name)
        self.breed = breed

    # Overriding the speak method
    def speak(self):
        print(f"{self.name} barks.")

# Creating objects of the classes
animal = Animal("Generic Animal")
dog = Dog("Buddy", "Golden Retriever")

# Calling methods
animal.speak()  # Output: Generic Animal makes a sound.
dog.speak()     # Output: Buddy barks.
````

````python
# Multiple Inheritance
class Animal:
    def __init__(self, name):
        self.name = name

class Pet:
    def __init__(self, owner):
        self.owner = owner

class Dog(Animal, Pet):
    def __init__(self, name, breed, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
        self.breed = breed

    def speak(self):
        print(f"{self.name} barks. Owned by {self.owner}.")

dog = Dog("Buddy", "Golden Retriever", "Alice")
dog.speak()  # Output: Buddy barks. Owned by Alice.
````

````python
# Multilevel Inheritance
class Animal:
    def speak(self):
        print("Animal makes a sound.")

class Mammal(Animal):
    def speak(self):
        print("Mammal makes a sound.")

class Dog(Mammal):
    def speak(self):
        print("Dog barks.")

dog = Dog()
dog.speak()  # Output: Dog barks.
````

In Python, the `super()` function is used to call methods from a **parent class** in the context of **inheritance**. It provides a way for a subclass to invoke methods or attributes of its parent class without explicitly referring to the parent class by name. This helps maintain a clean and flexible structure, especially in complex inheritance hierarchies.

**How `super()` Works**

1. **Calling Parent Class Methods:**
   The primary use of `super()` is to call a method from a parent class in a subclass. It allows you to extend or modify the behavior of the parent class method, while still leveraging its functionality.

2. **Accessing Parent Class Attributes:**
   You can also use `super()` to access attributes defined in the parent class.

3. **Method Resolution Order (MRO):**
   When using `super()`, Python uses the **Method Resolution Order (MRO)** to decide the order in which methods are called. This is particularly important in **multiple inheritance** where a class inherits from more than one parent class.

````python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")

class Dog(Animal):
    def __init__(self, name, breed):
        # Call the parent class constructor using super()
        super().__init__(name)
        self.breed = breed

    def speak(self):
        # Call the parent class speak method using super()
        super().speak()  # This invokes the 'speak' method from Animal
        print(f"{self.name} barks")

# Create an instance of Dog
dog = Dog("Buddy", "Golden Retriever")
dog.speak()
````

````python
# Using `super()` in Multiple Inheritance
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")

class Mammal(Animal):
    def speak(self):
        print(f"{self.name} makes a mammal sound")

class Dog(Mammal):
    def __init__(self, name, breed):
        # Call the parent classes' constructors
        super().__init__(name)

    def speak(self):
        super().speak()  # Calls speak from Mammal class (according to MRO)
        print(f"{self.name} barks")

# Creating an instance of Dog
dog = Dog("Buddy", "Golden Retriever")
dog.speak()
````

## Polymorphism

**Polymorphism** is a fundamental concept in object-oriented programming (OOP), which allows objects of different classes to be treated as objects of a common superclass. The term "polymorphism" comes from the Greek words "poly" (meaning many) and "morph" (meaning form), thus referring to "many forms." In simple terms, polymorphism allows one interface to be used for a general class of actions.

### Types of Polymorphism:
1. **Compile-time Polymorphism (Static Binding)**: This type of polymorphism is resolved at compile time. Method overloading and operator overloading are examples.
2. **Run-time Polymorphism (Dynamic Binding)**: This type of polymorphism is resolved at runtime. Method overriding is an example.

In Python, polymorphism is primarily **run-time polymorphism**, which occurs through **method overriding**.

````python
# Method Overriding
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class
class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Another child class
class Cat(Animal):
    def speak(self):
        print("Cat meows")

# Creating objects
animal = Animal()
dog = Dog()
cat = Cat()

# Calling the same method on different objects
animal.speak()  # Output: Animal makes a sound
dog.speak()     # Output: Dog barks
cat.speak()     # Output: Cat meows
````

````python
# With a Common Interface
class Shape:
    def area(self):
        pass  # To be implemented by subclasses

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2  # Area of a circle

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height  # Area of a rectangle

# Creating objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Using polymorphism
shapes = [circle, rectangle]

for shape in shapes:
    print(f"Area: {shape.area()}")  # Different areas are computed based on the object type
````

````python
# With Functions
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class
class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Another child class
class Cat(Animal):
    def speak(self):
        print("Cat meows")

# Creating objects
animal = Animal()
dog = Dog()
cat = Cat()

def make_animal_speak(animal):
    animal.speak()  # Calls the speak method, whatever the object type is

make_animal_speak(dog)  # Output: Dog barks
make_animal_speak(cat)  # Output: Cat meows
````

````python
# With super()
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        super().speak()  # Calls the method from Animal class
        print("Dog barks")

dog = Dog()
dog.speak()
````

# Decorators


A decorator is a function that takes another function (or method) as an argument, and returns a new function that usually extends or alters the behavior of the original function.

### Basic Syntax of a Decorator

The syntax for applying a decorator to a function is as follows:

```python
@decorator_function
def my_function():
    # original function code
```

This is equivalent to:

```python
def my_function():
    # original function code

my_function = decorator_function(my_function)
```


````python
# Basic Decorator
def my_decorator(func):
    def wrapper():
        print("Before the function is called.")
        func()
        print("After the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

# Call the decorated function
say_hello()
````

````python
# Using Decorators with Arguments
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before the function is called.")
        result = func(*args, **kwargs)  # Pass arguments to the original function
        print("After the function is called.")
        return result
    return wrapper

@my_decorator
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")
````

````python
# Nested Decorators
def decorator_1(func):
    def wrapper():
        print("Decorator 1")
        func()
    return wrapper

def decorator_2(func):
    def wrapper():
        print("Decorator 2")
        func()
    return wrapper

@decorator_1
@decorator_2
def greet():
    print("Hello!")

greet()
````

````python
# Decorators with Return values
def double_return_value(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result * 2
    return wrapper

@double_return_value
def add(a, b):
    return a + b

print(add(3, 4))  # Output: 14
````

````python
# Class-based Decorator
class MyDecorator:
    def __init__(self, func):
        self.func = func

    # The __call__ method allows an instance of the class to be used as a decorator.
    def __call__(self, *args, **kwargs):
        print("Before the function is called.")
        result = self.func(*args, **kwargs)
        print("After the function is called.")
        return result

@MyDecorator
def say_hello():
    print("Hello!")

say_hello()
````

#### `@throttle(2, 1)`  
Throttle function calls to 2 per second (example implementation of a throttle decorator).

```python
import time
from threading import Lock

def throttle(rate, period):
    lock = Lock()
    tokens = rate
    last_time = time.time()

    def decorator(func):
        def wrapper(*args, **kwargs):
            nonlocal tokens, last_time
            with lock:
                now = time.time()
                elapsed = now - last_time
                tokens = min(rate, tokens + elapsed * rate / period)
                if tokens >= 1:
                    tokens -= 1
                    last_time = now
                    return func(*args, **kwargs)
                else:
                    time.sleep(period / rate)
                    return wrapper(*args, **kwargs)
        return wrapper
    return decorator

@throttle(2, 1)
def print_message():
    print("Throttled message at", time.time())

for _ in range(5):
    print_message()
```

# Generators

### **1. Iterable**

An **iterable** is any Python object capable of returning its elements one at a time. It must implement the `__iter__()` method (which returns an iterator) or the `__getitem__()` method to access elements.

- **Examples of Iterables:** Lists, tuples, strings, dictionaries, sets, ranges, and more.
- **Key Characteristics:**
  - Can be iterated over using a loop (e.g., `for` loop).
  - Does not directly maintain state.

#### Example:
```python
my_list = [1, 2, 3]  # List is an iterable
for item in my_list:
    print(item)

# Under the hood
iterator = iter(my_list)  # Gets an iterator from the iterable
print(next(iterator))  # 1
print(next(iterator))  # 2
```

---

### **2. Iterator**

An **iterator** is an object representing a stream of data. It is created by calling the `iter()` method on an iterable and provides values one at a time. An iterator must implement two methods:

1. **`__iter__()`**: Returns the iterator object itself.
2. **`__next__()`**: Returns the next value from the iterator. Raises `StopIteration` when there are no more items.

- **Key Characteristics:**
  - Maintains state: Keeps track of where it is during iteration.
  - Can be iterated only once (consumable).

#### Example:
```python
my_list = [1, 2, 3]  # List is an iterable
iterator = iter(my_list)  # Convert iterable to an iterator

print(next(iterator))  # 1
print(next(iterator))  # 2
print(next(iterator))  # 3
# print(next(iterator))  # Raises StopIteration
```

---

### **3. Generator**

A **generator** is a special type of iterator created using either a **generator function** (with the `yield` keyword) or a **generator expression**. It lazily produces values on demand, which makes it memory-efficient.

- **Key Characteristics:**
  - Implements both `__iter__()` and `__next__()`.
  - Can produce values lazily (on-demand).
  - Often more compact and convenient than manually implementing iterators.

#### Example:
1. **Using a Generator Function:**
   ```python
   def my_generator(count=0):
      while True:
         yield count
         count += 1

   gen = my_generator()
   print(next(gen))  # 1
   print(next(gen))  # 2
   ```

2. **Using a Generator Expression:**
   ```python
   gen = (x * x for x in range(3))
   print(next(gen))  # 0
   print(next(gen))  # 1
   ```

---

### **Comparison Table**

| **Feature**          | **Iterable**                         | **Iterator**                     | **Generator**                                     |
|-----------------------|--------------------------------------|-----------------------------------|-------------------------------------------------|
| **Definition**        | An object that can return an iterator. | An object that produces values one at a time. | A special type of iterator defined with `yield`. |
| **Examples**          | List, tuple, string, dictionary.     | Result of `iter(iterable)`.       | Created using generator functions or expressions. |
| **Implements**        | `__iter__()` or `__getitem__()`.     | `__iter__()` and `__next__()`.   | `__iter__()` and `__next__()`.                  |
| **State Maintenance** | No.                                  | Yes.                              | Yes.                                            |
| **Consumption**       | Can be reused.                      | One-time use.                     | One-time use.                                   |
| **Lazy Evaluation**   | No.                                  | Depends on implementation.        | Yes.                                            |
| **Memory Efficiency** | Uses memory for all elements.        | Efficient (depends on implementation). | Highly efficient due to lazy evaluation.        |

## Generator Methods

Generators in Python have several specialized methods that allow for interaction and control beyond simple iteration. These methods include:

---

### **1. `next()`**
- **Purpose:** Retrieves the next value from the generator.
- **Behavior:** If there are no more values to yield, raises a `StopIteration` exception.
- **Usage:**
  ```python
  def my_gen():
      yield 1
      yield 2

  gen = my_gen()
  print(next(gen))  # 1
  print(next(gen))  # 2
  # print(next(gen))  # Raises StopIteration
  ```

---

### **2. `send(value)`**
- **Purpose:** Resumes the generator's execution and sends a value into the generator, which is received where it is paused.
- **Behavior:**
  - The first call to `send()` must be preceded by `next()` or equivalent to advance the generator to the first `yield`.
  - The `send()` method sends a value that becomes the result of the `yield` expression where the generator is paused.
- **Usage:**
  ```python
  def my_gen():
      value = yield "Start"
      yield f"Received: {value}"

  gen = my_gen()
  print(next(gen))       # "Start"
  print(gen.send("Hi"))  # "Received: Hi"
  ```

---

### **3. `throw(exception_type, value=None, traceback=None)`**
- **Purpose:** Raises an exception at the point where the generator is paused.
- **Behavior:** The generator can handle the exception with a `try`/`except` block or propagate it further.
- **Usage:**
  ```python
  def my_gen():
      try:
          yield "Start"
      except ValueError:
          yield "Exception caught"

  gen = my_gen()
  print(next(gen))           # "Start"
  print(gen.throw(ValueError))  # "Exception caught"
  ```

---

### **4. `close()`**
- **Purpose:** Stops the generator by raising a `GeneratorExit` exception at the point where it is paused.
- **Behavior:**
  - If the generator handles `GeneratorExit`, it should terminate the execution.
  - After `close()`, further calls to `next()` or `send()` raise `StopIteration`.
- **Usage:**
  ```python
  def my_gen():
      try:
          yield "Start"
      except GeneratorExit:
          print("Generator closed")

  gen = my_gen()
  print(next(gen))  # "Start"
  gen.close()       # "Generator closed"
  ```

---

### **Complete Example Using All Methods**

```python
def controlled_generator():
    print("Generator started")
    try:
        value = yield "Ready"
        while True:
            print(f"Received: {value}")
            value = yield f"Processed: {value}"
    except ValueError:
        yield "ValueError caught!"
    except GeneratorExit:
        print("Generator shutting down")

# Create generator
gen = controlled_generator()

# Advance to first yield
print(next(gen))  # "Generator started" -> "Ready"

# Send a value
print(gen.send(42))  # "Received: 42" -> "Processed: 42"

# Handle exception
print(gen.throw(ValueError))  # "ValueError caught!"

# Close the generator
gen.close()  # "Generator shutting down"
```

Output:
```
Generator started
Ready
Received: 42
Processed: 42
ValueError caught!
Generator shutting down
```

### **Summary of Generator Methods**

## Class-based Generators

```python
class FibonacciGenerator:
    def __init__(self, limit):
        self.limit = limit
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.a > self.limit:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        return value

    def reset(self):
        """Reset the generator to start from the beginning."""
        self.a, self.b = 0, 1

# Using the generator
fib = FibonacciGenerator(100)
for number in fib:
    print(number)

fib.reset()
print("After reset:")
for number in fib:
    print(number)
```

````python
class FibonacciGenerator:
    def __init__(self, limit):
        self.limit = limit
        self.sequence = []
        self._generate_sequence()

    def _generate_sequence(self):
        """Generate the Fibonacci sequence up to the limit."""
        a, b = 0, 1
        while a <= self.limit:
            self.sequence.append(a)
            a, b = b, a + b

    def __iter__(self):
        """Make the generator iterable."""
        self.index = 0  # Reset index for iteration
        return self

    def __next__(self):
        """Provide the next value during iteration."""
        if self.index >= len(self.sequence):
            raise StopIteration
        value = self.sequence[self.index]
        self.index += 1
        return value

    def __getitem__(self, index):
        """Allow indexed access to the Fibonacci sequence."""
        if index < 0 or index >= len(self.sequence):
            raise IndexError("Index out of range")
        return self.sequence[index]

    def __len__(self):
        """Return the length of the Fibonacci sequence."""
        return len(self.sequence)

    def __contains__(self, value):
        """Check if a value exists in the Fibonacci sequence."""
        return value in self.sequence

    def __repr__(self):
        """Provide a string representation of the generator."""
        return f"FibonacciGenerator(limit={self.limit}, sequence={self.sequence})"


# Using the FibonacciGenerator class
fib = FibonacciGenerator(50)

# Iteration
print("Iteration:")
for number in fib:
    print(number)

# Indexing
print("\nIndexing:")
print(f"Element at index 5: {fib[5]}")

# Length
print("\nLength:")
print(f"Length of the sequence: {len(fib)}")

# Membership
print("\nMembership:")
print(f"Is 21 in the sequence? {'Yes' if 21 in fib else 'No'}")
print(f"Is 22 in the sequence? {'Yes' if 22 in fib else 'No'}")

# Representation
print("\nRepresentation:")
print(fib)
````

# Regular Expressions

### **1. Regular Expression Basics**

| **Symbol** | **Description**                                     | **Example**                     |
|------------|-----------------------------------------------------|----------------------------------|
| `.`        | Matches any character except newline.               | `a.b` matches `acb`, `a1b`.     |
| `^`        | Matches the start of the string.                    | `^hello` matches `hello world`. |
| `$`        | Matches the end of the string.                      | `world$` matches `hello world`. |
| `*`        | Matches 0 or more repetitions of the preceding.     | `a*` matches `aa`, `a`, or ``.  |
| `+`        | Matches 1 or more repetitions of the preceding.     | `a+` matches `a`, `aa`.         |
| `?`        | Matches 0 or 1 repetition of the preceding.         | `ab?` matches `a` or `ab`.      |
| `{n}`      | Matches exactly `n` repetitions.                    | `a{3}` matches `aaa`.           |
| `{n,}`     | Matches `n` or more repetitions.                    | `a{2,}` matches `aa`, `aaa`.    |
| `{n,m}`    | Matches between `n` and `m` repetitions.            | `a{2,4}` matches `aa`, `aaa`.   |
| `\|`       | Logical OR (alternation).                           | `cat\|dog` matches `cat` or `dog`. |

---

### **2. Character Classes**

| **Symbol**     | **Description**                                             | **Example**                        |
|-----------------|-------------------------------------------------------------|-------------------------------------|
| `[abc]`         | Matches any one of `a`, `b`, or `c`.                        | `[aeiou]` matches any vowel.       |
| `[^abc]`        | Matches anything except `a`, `b`, or `c`.                   | `[^aeiou]` matches non-vowels.     |
| `[a-z]`         | Matches any character from `a` to `z`.                      | `[0-9]` matches digits.            |
| `[a-zA-Z0-9]`   | Matches letters and digits.                                 | `[a-zA-Z]` matches letters only.   |
| `\d`            | Matches any digit (equivalent to `[0-9]`).                  | `\d{3}` matches `123`.             |
| `\D`            | Matches any non-digit (equivalent to `[^0-9]`).             | `\D` matches `abc`.                |
| `\w`            | Matches any alphanumeric character or underscore (`_`).     | `\w+` matches `word123`.           |
| `\W`            | Matches any non-alphanumeric character.                     | `\W` matches `@!`.                 |
| `\s`            | Matches any whitespace character (`\t`, `\n`, space).       | `\s+` matches spaces or tabs.      |
| `\S`            | Matches any non-whitespace character.                       | `\S` matches `word`.               |

---

### **3. Anchors and Boundaries**

| **Symbol** | **Description**                                               | **Example**                       |
|------------|---------------------------------------------------------------|------------------------------------|
| `^`        | Matches the start of the string or line (in multiline mode).  | `^Hello` matches `Hello world`.   |
| `$`        | Matches the end of the string or line (in multiline mode).    | `world$` matches `hello world`.   |
| `\b`       | Matches a word boundary.                                      | `\bword\b` matches `word` only.   |
| `\B`       | Matches a non-word boundary.                                  | `\Bword` matches `password`.      |

---

### **4. Special Sequences**

| **Symbol**  | **Description**                                 | **Example**                      |
|-------------|-------------------------------------------------|-----------------------------------|
| `\A`        | Matches the start of the string.               | `\AHello` matches `Hello world`. |
| `\Z`        | Matches the end of the string.                 | `world\Z` matches `hello world`. |
| `\G`        | Matches where the last match ended.            | Not commonly used in Python.     |
| `(?=...)`   | Positive lookahead (matches if `...` follows). | `\w+(?=\s)` matches `word` in `word `. |
| `(?!...)`   | Negative lookahead (matches if `...` does not follow). | `word(?!\s)` excludes trailing spaces. |
| `(?<=...)`  | Positive lookbehind (matches if `...` precedes). | `(?<=\$)\d+` matches `123` in `$123`. |
| `(?<!...)`  | Negative lookbehind (matches if `...` does not precede). | `(?<!\$)\d+` matches numbers not preceded by `$`. |

---

### **5. Groups and Capturing**

| **Syntax**     | **Description**                                                   | **Example**                         |
|-----------------|-------------------------------------------------------------------|--------------------------------------|
| `(...)`         | Captures the matched substring for backreferences.               | `(a|b)c` matches `ac` or `bc`.      |
| `(?:...)`       | Non-capturing group (does not save match for backreferences).     | `(?:a|b)c` matches `ac` or `bc`.    |
| `(?P<name>...)` | Named capturing group.                                           | `(?P<area>\d{3})` names the group.  |
| `\number`       | Refers to a captured group by number.                            | `(\d)\1` matches `11`, `22`.        |
| `(?P=name)`     | Refers to a named group.                                         | `(?P<area>\d{3})-\g<area>` matches `123-123`. |

---

### **6. Flags in Regular Expressions**

| **Flag**         | **Description**                                     | **Usage**                 |
|-------------------|-----------------------------------------------------|---------------------------|
| `re.IGNORECASE`   | Makes matching case-insensitive.                    | `re.search(r'abc', text, re.I)` |
| `re.MULTILINE`    | Allows `^` and `$` to match the start and end of lines. | `re.search(r'^abc', text, re.M)` |
| `re.DOTALL`       | Makes `.` match newline characters.                 | `re.search(r'.+', text, re.S)` |
| `re.VERBOSE`      | Allows the regex to be written in a more readable form. | `re.search(r"abc  # comment", re.X)` |
| `re.ASCII`        | Makes `\w`, `\W`, `\b`, `\B` only match ASCII chars. | `re.search(r'\w+', text, re.A)` |

---

### **7. Functions in `re` Module**

| **Function**      | **Description**                                             | **Example**                             |
|--------------------|-------------------------------------------------------------|------------------------------------------|
| `re.match()`       | Matches the pattern at the start of the string.             | `re.match(r'\d+', '123abc')` matches `123`. |
| `re.search()`      | Searches for the first match anywhere in the string.        | `re.search(r'\d+', 'abc123')` matches `123`. |
| `re.findall()`     | Returns a list of all matches.                              | `re.findall(r'\d+', 'abc123def456')` gives `['123', '456']`. |
| `re.finditer()`    | Returns an iterator of match objects.                       | `re.finditer(r'\d+', '123 456')`.        |
| `re.sub()`         | Replaces matches with a specified string.                   | `re.sub(r'\d+', '#', '123abc456')`.      |
| `re.subn()`        | Like `re.sub()`, but also returns the number of substitutions. | `re.subn(r'\d+', '#', '123abc456')`.    |
| `re.split()`       | Splits the string by the occurrences of the pattern.        | `re.split(r',\s*', 'a, b, c')`.          |
| `re.compile()`     | Compiles a regex pattern for reuse.                         | `pattern = re.compile(r'\d+')`.          |

## `shutil` module

The **`shutil`** module in Python provides a higher-level interface for file operations, such as copying, moving, and removing files and directories. It is often used in scenarios requiring more complex file or directory management than what the **`os`** module offers.

---

### Key Features of the `shutil` Module

#### **1. File Operations**
- **`shutil.copy(src, dst)`**: Copies a file from `src` to `dst`. If `dst` is a directory, the file is copied into it with the same name.
- **`shutil.copy2(src, dst)`**: Copies a file, preserving metadata (e.g., modification times).
- **`shutil.copyfile(src, dst)`**: Copies the content of `src` to `dst`. Both `src` and `dst` must be file paths.
- **`shutil.copymode(src, dst)`**: Copies the permissions of `src` to `dst`.
- **`shutil.copystat(src, dst)`**: Copies the metadata of `src` to `dst`.

**Example:**
```python
import shutil

shutil.copy('source.txt', 'destination.txt')  # Copy file
shutil.copy2('source.txt', '/backup/')       # Copy with metadata
```

---

#### **2. Directory Operations**
- **`shutil.copytree(src, dst, dirs_exist_ok=False)`**: Recursively copies a directory tree from `src` to `dst`.
  - `dirs_exist_ok=True` allows existing directories in the destination.
- **`shutil.rmtree(path)`**: Recursively deletes a directory tree.

**Example:**
```python
import shutil

shutil.copytree('source_folder', 'backup_folder')  # Copy entire directory
shutil.rmtree('backup_folder')                    # Delete directory
```

---

#### **3. Moving and Renaming**
- **`shutil.move(src, dst)`**: Moves a file or directory to the destination. Can also be used to rename files or directories.

**Example:**
```python
import shutil

shutil.move('old_name.txt', 'new_name.txt')  # Rename file
shutil.move('file.txt', '/new_location/')   # Move file
```

---

#### **4. Archiving**
- **`shutil.make_archive(base_name, format, root_dir)`**: Creates an archive (e.g., `.zip`, `.tar`) from the specified directory.
  - `format` can be `'zip'`, `'tar'`, `'gztar'`, `'bztar'`, or `'xztar'`.

- **`shutil.unpack_archive(filename, extract_dir=None, format=None)`**: Extracts an archive to the specified directory.

**Example:**
```python
import shutil

# Create an archive
shutil.make_archive('backup', 'zip', 'source_folder')

# Extract an archive
shutil.unpack_archive('backup.zip', 'extracted_folder')
```

---

#### **5. Disk Usage**
- **`shutil.disk_usage(path)`**: Returns the disk usage statistics for the given path as a named tuple: `(total, used, free)`.

**Example:**
```python
import shutil

usage = shutil.disk_usage('/')
print(f"Total: {usage.total // (1024**3)} GB")
print(f"Used: {usage.used // (1024**3)} GB")
print(f"Free: {usage.free // (1024**3)} GB")
```

---

#### **6. Advanced File Matching**
- **`shutil.which(cmd)`**: Checks if a command is available in the system's PATH. Useful for verifying the existence of executables.

**Example:**
```python
import shutil

cmd_path = shutil.which('python')
if cmd_path:
    print(f"Python found at: {cmd_path}")
else:
    print("Python is not in PATH")
```

---

### Use Cases of the `shutil` Module
1. **Backup and Restore:**
   - Copy files or directories to create backups.
   - Create or extract archives for deployment or distribution.

2. **File Management:**
   - Automate file and directory movement.
   - Manage temporary files during a program's execution.

3. **Disk Management:**
   - Check available disk space for tasks like data processing.

4. **File Permission Management:**
   - Copy or replicate file permissions and metadata.

---

### Example: Backup and Restore Script
```python

# Create a backup
shutil.copytree('project', 'project_backup')

# Create a ZIP archive of the backup
shutil.make_archive('project_backup', 'zip', 'project_backup')

# Clean up the backup directory

print("Backup and archiving complete!")
```

The `shutil` module is powerful and simplifies many file and directory operations, making it a go-to tool for tasks involving file management.

## `logging` module

The **`logging`** module in Python provides a flexible framework for emitting log messages from Python programs. It is a standard way to record events, errors, and other information for debugging or tracking the execution of applications.

### Why Use the `logging` Module?

- **Centralized Logging**: Avoid using `print()` for debugging and manage logs systematically.
- **Configurable Levels**: Control the verbosity of logs (e.g., debug, info, error).
- **Multiple Outputs**: Send logs to files, consoles, or external systems.
- **Thread-Safe**: Can be used in multi-threaded applications.

### Logging Levels

The `logging` module defines the following standard levels for logging, in increasing order of severity:

| Level       | Numeric Value | Description                                      |
|-------------|---------------|--------------------------------------------------|
| `DEBUG`     | 10            | Detailed information, useful for debugging.      |
| `INFO`      | 20            | General information about program execution.     |
| `WARNING`   | 30            | Indicates a potential problem.                   |
| `ERROR`     | 40            | A serious error occurred, unable to proceed.     |
| `CRITICAL`  | 50            | A very serious error, program may crash.         |

### Basic Usage

4. **Default Values**: Assign default values to arguments if none are provided.

---

### Basic Workflow

1. Create an `ArgumentParser` object.
2. Define arguments and options using `add_argument`.
3. Parse arguments with `parse_args`.
4. Access the parsed arguments as attributes of the returned object.

---

### Example: Basic Usage

## `random` module

The **`random`** module in Python provides functions for generating random numbers, selecting random elements, and working with random distributions. It is useful for tasks like simulations, randomized testing, and creating random samples.

---

### Key Features of the `random` Module

#### **1. Random Number Generation**
- **`random.random()`**: Generates a random float between 0.0 (inclusive) and 1.0 (exclusive).
- **`random.randint(a, b)`**: Returns a random integer \( N \) such that \( a \leq N \leq b \).
- **`random.randrange(start, stop[, step])`**: Returns a random number from a range, similar to `range()`.
- **`random.uniform(a, b)`**: Returns a random float \( N \) such that \( a \leq N \leq b \).

**Example:**
```python
import random

print(random.random())        # 0.37444887175646646 (example)
print(random.randint(1, 10))  # Random integer between 1 and 10
print(random.randrange(0, 20, 2))  # Random even number between 0 and 20
print(random.uniform(1.5, 3.5))    # Random float between 1.5 and 3.5
```

---

#### **2. Random Choices**
- **`random.choice(seq)`**: Selects a random element from a non-empty sequence.
- **`random.choices(population, weights=None, k=1)`**: Returns a list of `k` random elements from `population`, with optional weighting.
- **`random.sample(population, k)`**: Returns a list of `k` unique elements from `population`.

**Example:**
```python
import random

items = ['apple', 'banana', 'cherry', 'date']

print(random.choice(items))              # Randomly select one item
print(random.choices(items, k=2))        # Select 2 items (with replacement)
print(random.sample(items, 3))           # Select 3 items (without replacement)
```

---

#### **3. Shuffling**
- **`random.shuffle(seq)`**: Shuffles the sequence in place.

**Example:**
```python
import random

deck = ['Ace', 'King', 'Queen', 'Jack']
random.shuffle(deck)
print(deck)  # ['Queen', 'Jack', 'Ace', 'King'] (example)
```

---

#### **4. Random Distributions**
- **`random.gauss(mu, sigma)`**: Returns a random value from a Gaussian (normal) distribution.
- **`random.expovariate(lambd)`**: Returns a random value from an exponential distribution.
- **`random.uniform(a, b)`**: Returns a random float uniformly distributed between \( a \) and \( b \).
- **`random.triangular(low, high, mode)`**: Returns a random value from a triangular distribution.

**Example:**
```python
import random

print(random.gauss(0, 1))           # Gaussian distribution with mean 0, std dev 1
print(random.expovariate(1/10))     # Exponential distribution with lambda 0.1
print(random.triangular(1, 10, 5))  # Triangular distribution with mode at 5
```

---

## `collections` module

The **`collections`** module in Python provides specialized container datatypes that extend the functionality of Python's built-in data structures (like lists, tuples, and dictionaries). It is particularly useful for tasks requiring enhanced or alternative data manipulation.

---

### Key Data Structures in `collections`

#### **1. Counter**
- A dictionary subclass for counting hashable objects.
- Elements are stored as dictionary keys, and their counts are stored as values.

**Usage:**
```python
from collections import Counter

# Example: Counting occurrences
data = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
counter = Counter(data)
print(counter)  # Counter({'apple': 3, 'banana': 2, 'orange': 1})

# Access counts
print(counter['apple'])  # 3

# Most common elements
print(counter.most_common(2))  # [('apple', 3), ('banana', 2)]

# Update counts
counter.update(['banana', 'orange', 'apple'])
print(counter)  # Counter({'apple': 4, 'banana': 3, 'orange': 2})
```

---

#### **2. defaultdict**
- A dictionary subclass that provides default values for keys that are missing.

**Usage:**
```python
from collections import defaultdict

# Example: Grouping items
dd = defaultdict(list)
data = [('fruit', 'apple'), ('fruit', 'banana'), ('vegetable', 'carrot')]
for key, value in data:
    dd[key].append(value)

print(dd)  # defaultdict(<class 'list'>, {'fruit': ['apple', 'banana'], 'vegetable': ['carrot']})
```

**Common Default Factories:**
- `list`
- `int` (default 0)
- `set`

---

#### **3. OrderedDict** (Python 3.7+ dictionaries are already ordered)
- A dictionary subclass that maintains the insertion order of items.

**Usage:**
```python
from collections import OrderedDict

od = OrderedDict()
od['a'] = 1
od['b'] = 2
od['c'] = 3
print(od)  # OrderedDict([('a', 1), ('b', 2), ('c', 3)])
```

**Special Methods:**
- `move_to_end(key, last=True)`: Move an existing key to the end or the beginning.
- Reversible with `reversed(od)`.

---

#### **4. namedtuple**
- A factory function for creating tuple subclasses with named fields.
- Provides a readable and immutable alternative to regular tuples.

**Usage:**
```python
from collections import namedtuple

# Define a namedtuple
Point = namedtuple('Point', ['x', 'y'])

p = Point(10, 20)
print(p.x, p.y)  # 10, 20
print(p)         # Point(x=10, y=20)

# Namedtuples are immutable
# p.x = 30  # Raises AttributeError
```

---

#### **5. deque** (Double-Ended Queue)
- A general-purpose, thread-safe, and memory-efficient queue implementation.
- Faster appends and pops from both ends compared to lists.

**Usage:**
```python
from collections import deque

# Create a deque
dq = deque([1, 2, 3])

# Add elements
dq.append(4)        # Add to the right
dq.appendleft(0)    # Add to the left
print(dq)           # deque([0, 1, 2, 3, 4])

# Remove elements
dq.pop()            # Remove from the right
dq.popleft()        # Remove from the left
print(dq)           # deque([1, 2, 3])

# Rotate elements
dq.rotate(1)        # Rotate right
print(dq)           # deque([3, 1, 2])
dq.rotate(-1)       # Rotate left
print(dq)           # deque([1, 2, 3])
```

---

#### **6. ChainMap**
- Combines multiple dictionaries into a single view.

**Usage:**
```python
from collections import ChainMap

dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
chain = ChainMap(dict1, dict2)
print(chain['a'])  # 1 (from dict1)
print(chain['b'])  # 2 (from dict1)
print(chain['c'])  # 4 (from dict2)
```

---

#### **7. UserDict, UserList, UserString**
- Wrapper classes for creating dictionary-like, list-like, or string-like objects.
- Useful for subclassing and customization.

**Usage:**
```python
from collections import UserDict

class MyDict(UserDict):
    def __setitem__(self, key, value):
        if not isinstance(key, str):
            raise ValueError("Keys must be strings!")
        super().__setitem__(key, value)

d = MyDict()
d['key'] = 'value'
print(d)  # {'key': 'value'}
# d[1] = 'value'  # Raises ValueError
```

| **Type**         | **Purpose**                                                                 |
|-------------------|-----------------------------------------------------------------------------|

| `OrderedDict`     | Preserving the order of insertion.                                         |
| `namedtuple`      | Readable, immutable tuples with named fields.                             |
| `deque`           | Optimized list-like container with fast appends/pops from both ends.       |
| `ChainMap`        | Combining multiple dictionaries into a single view.                       |
| `UserDict`        | Extending and customizing dictionary-like objects.                        |
| `UserList`        | Extending and customizing list-like objects.                              |

### Example: Combining Features

## `itertools` module

The **`itertools`** module in Python provides a collection of tools for creating and working with iterators. It includes functions for efficient looping, combinatorics, and creating specialized iterators.

---

### Key Features of the `itertools` Module

The module is divided into three main categories:

1. **Infinite Iterators**: Generate infinite sequences.
2. **Iterators Terminating on Input**: Produce finite sequences based on input data.
3. **Combinatoric Iterators**: Generate permutations, combinations, and Cartesian products.

---

### **1. Infinite Iterators**

#### **`itertools.count(start=0, step=1)`**
Generates an infinite sequence starting from `start`, incremented by `step`.

**Usage:**
```python
import itertools

for i in itertools.count(10, 2):  # Starts at 10, steps by 2
    if i > 20:
        break
    print(i)  # 10, 12, 14, 16, 18, 20
```

---

#### **`itertools.cycle(iterable)`**
Repeats elements of an iterable infinitely.

**Usage:**
```python
import itertools

colors = ['red', 'green', 'blue']
for i, color in zip(range(10), itertools.cycle(colors)):
    print(color)
```

---

#### **`itertools.repeat(object, times=None)`**
Repeats an object a specified number of times (or infinitely if `times` is not specified).

**Usage:**
```python

for item in itertools.repeat('A', 3):
    print(item)  # A, A, A
```

---

### **2. Iterators Terminating on Input**

#### **`itertools.accumulate(iterable, func=operator.add)`**
Returns accumulated sums (or other binary operations specified by `func`).

**Usage:**
```python
import itertools

print(list(itertools.accumulate(nums)))  # [1, 3, 6, 10]

import operator
print(list(itertools.accumulate(nums, operator.mul)))  # [1, 2, 6, 24]
```

#### **`itertools.chain(*iterables)`**
Chains multiple iterables into a single iterator.

**Usage:**

#### **`itertools.compress(data, selectors)`**
Filters elements from `data` based on `selectors`.

**Usage:**
```python
import itertools

data = ['a', 'b', 'c', 'd']
selectors = [1, 0, 1, 0]

```

---

#### **`itertools.dropwhile(predicate, iterable)`**
Drops elements from the iterable as long as the predicate is `True`; rest are returned.

**Usage:**

```python
import itertools

nums = [1, 2, 3, 4, 5]
```

---

#### **`itertools.takewhile(predicate, iterable)`**
Returns elements from the iterable as long as the predicate is `True`.

**Usage:**
```python
import itertools

nums = [1, 2, 3, 4, 5]

```

#### **`itertools.islice(iterable, start, stop, step)`**
Returns selected elements from the iterable, slicing it like a list.

**Usage:**
```python
import itertools

nums = range(10)
print(list(itertools.islice(nums, 2, 8, 2)))  # [2, 4, 6]
```

#### **`itertools.starmap(function, iterable)`**
Applies a function to the elements of the iterable (unpacking arguments).

**Usage:**
```python

print(list(itertools.starmap(pow, data)))  # [32, 9, 1000]
```

### **3. Combinatoric Iterators**

#### **`itertools.product(*iterables, repeat=1)`**
Returns the Cartesian product of input iterables.

**Usage:**

#### **`itertools.permutations(iterable, r=None)`**

#### **`itertools.combinations(iterable, r)`**

## `functools` module

The **`functools`** module in Python provides higher-order functions that act on or return other functions. It is particularly useful for functional programming tasks, such as creating decorators, partial functions, and improving performance with caching.

---

### Key Features of the `functools` Module

#### **1. `functools.lru_cache`**
- A decorator for caching the results of expensive function calls.
- Stands for **Least Recently Used** cache.

**Usage:**
```python
import functools

@functools.lru_cache(maxsize=128)  # Cache up to 128 calls
def fib(n):
    if n < 2:
        return n
    return fib(n - 1) + fib(n - 2)

print(fib(10))  # 55
print(fib.cache_info())  # Shows cache statistics
```

**Key Methods:**
- `fib.cache_clear()`: Clears the cache.
- `fib.cache_info()`: Returns cache statistics.

---

#### **2. `functools.partial`**
- Allows you to create a new function by fixing some arguments of an existing function.

**Usage:**
```python
import functools

def multiply(x, y):
    return x * y

# Create a new function with `y` fixed to 10
multiply_by_10 = functools.partial(multiply, y=10)
print(multiply_by_10(5))  # 50
```

---

#### **3. `functools.reduce`**
- Applies a function cumulatively to the items in a sequence, reducing the sequence to a single value.

**Usage:**
```python
import functools

nums = [1, 2, 3, 4]
result = functools.reduce(lambda x, y: x * y, nums)
print(result)  # 24
```

---

#### **4. `functools.cached_property`** *(Python 3.8+)*
- A decorator that converts a method into a property whose value is computed once and then cached.

**Usage:**
```python
import functools

class Circle:
    def __init__(self, radius):
        self.radius = radius

    @functools.cached_property
    def area(self):
        print("Calculating area...")
        return 3.14159 * self.radius ** 2

circle = Circle(10)
print(circle.area)  # Calculates and caches the area
print(circle.area)  # Uses the cached value
```

---

#### **5. `functools.total_ordering`**
- A class decorator that simplifies implementing comparison methods.

**Usage:**
```python
import functools

@functools.total_ordering
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __eq__(self, other):
        return self.age == other.age

    def __lt__(self, other):
        return self.age < other.age

p1 = Person("Alice", 30)
p2 = Person("Bob", 25)

print(p1 > p2)  # True
```

---

#### **6. `functools.singledispatch`**
- A decorator for creating single-dispatch generic functions (functions that behave differently based on the type of their first argument).

**Usage:**
```python
import functools

@functools.singledispatch
def process_data(data):
    raise NotImplementedError("Unsupported type")

@process_data.register(int)
def _(data):
    return f"Processing integer: {data}"

@process_data.register(str)
def _(data):
    return f"Processing string: {data}"

print(process_data(42))   # Processing integer: 42
print(process_data("Hi")) # Processing string: Hi
```

---

#### **7. `functools.wraps`**
- A decorator for preserving the metadata of the original function when writing custom decorators.

**Usage:**
```python
import functools

def my_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print("Before the function call")
        result = func(*args, **kwargs)
        print("After the function call")
        return result
    return wrapper

@my_decorator
def greet(name):
    """Greet a person."""
    print(f"Hello, {name}!")

print(greet.__name__)  # greet (preserved by functools.wraps)
print(greet.__doc__)   # Greet a person. (preserved by functools.wraps)
```

---

### Common `functools` Functions Summary

| **Function**           | **Description**                                                                 |
|-------------------------|---------------------------------------------------------------------------------|
| `lru_cache`            | Adds memoization (caching) to a function to improve performance.                |
| `partial`              | Creates a new function with some arguments fixed.                              |
| `reduce`               | Reduces a sequence to a single value by applying a function cumulatively.       |
| `cached_property`      | Creates a property that is computed once and cached for subsequent accesses.    |
| `total_ordering`       | Simplifies implementing comparison methods for a class.                        |
| `singledispatch`       | Enables creating type-specific behavior for a generic function.                |
| `wraps`                | Preserves metadata of original functions when writing decorators.              |

---

### Example: Combining Features

```python
import functools

@functools.lru_cache(maxsize=None)
def factorial(n):
    """Compute the factorial of n."""
    if n == 0:
        return 1
    return n * factorial(n - 1)

@functools.singledispatch
def describe(value):
    return f"Unknown type: {type(value)}"

@describe.register(int)
def _(value):
    return f"Integer: {value}"

@describe.register(str)
def _(value):
    return f"String: {value}"

print(factorial(5))  # 120
print(describe(10))  # Integer: 10
print(describe("Hi"))  # String: Hi
```

---

### When to Use `functools`

- **Performance**: Use `lru_cache` for optimizing expensive functions.
- **Simplification**: Use `partial` to predefine arguments for reuse.
- **Custom Decorators**: Use `wraps` to maintain metadata.
- **Flexible Type Handling**: Use `singledispatch` for generic type-specific behavior.
- **Data Integrity**: Use `total_ordering` to simplify comparisons in classes.

The `functools` module is a must-have tool for enhancing the efficiency, flexibility, and readability of Python code.

#### **Custom Deserialization**

```python
import json

    def __init__(self, name, date):
        self.name = name
        self.date = date

def custom_decoder(dct):

        dct["date"] = datetime.fromisoformat(dct["date"])
        return Event(dct["name"], dct["date"])
    return dct

json_string = '{"name": "Conference", "date": "2025-01-14T12:00:00"}'
event = json.loads(json_string, object_hook=custom_decoder)
print(event.name, event.date)
```

## `time` module

The **`time`** module in Python provides functions for working with time-related tasks, such as retrieving the current time, measuring elapsed time, and working with delays. It operates at a lower level compared to the `datetime` module, focusing on epoch time (seconds since January 1, 1970) and delays.

---

### Key Features of the `time` Module

1. **Epoch Time**: Represents time as the number of seconds since January 1, 1970 (UTC).
2. **Delays**: Introduce pauses in program execution.
3. **Time Measurements**: Measure performance and elapsed time.
4. **Formatting**: Convert between epoch time and human-readable formats.

---

### Common Functions in the `time` Module

#### **1. `time.time()`**
- Returns the current time in seconds since the epoch as a floating-point number.

**Usage:**
```python
import time

current_time = time.time()
print("Current Epoch Time:", current_time)
```

---

#### **2. `time.sleep(seconds)`**
- Suspends (pauses) the program for the specified number of seconds.

**Usage:**
```python
import time

print("Starting...")
time.sleep(2)  # Pause for 2 seconds
print("2 seconds later...")
```

---

#### **3. `time.ctime(seconds=None)`**
- Converts epoch time into a human-readable string.
- If `seconds` is not provided, the current time is used.

**Usage:**
```python
import time

print("Current Time:", time.ctime())  # Current time
print("Specific Time:", time.ctime(1672531200))  # Specific epoch time
```

---

#### **4. `time.localtime(seconds=None)`**
- Converts epoch time to a `struct_time` object in local time.
- If `seconds` is not provided, the current time is used.

**Usage:**
```python
import time

local_time = time.localtime()
print("Local Time:", local_time)
print("Year:", local_time.tm_year)
```

---

#### **5. `time.gmtime(seconds=None)`**
- Converts epoch time to a `struct_time` object in UTC (Greenwich Mean Time).
- If `seconds` is not provided, the current time is used.

**Usage:**
```python
import time

utc_time = time.gmtime()
print("UTC Time:", utc_time)
```

---

#### **6. `time.strftime(format, t)`**
- Formats a `struct_time` object into a string based on the given format.

**Usage:**
```python
import time

local_time = time.localtime()
formatted_time = time.strftime("%Y-%m-%d %H:%M:%S", local_time)
print("Formatted Time:", formatted_time)
```

---

#### **7. `time.strptime(string, format)`**
- Parses a string representing time into a `struct_time` object.

**Usage:**
```python
import time

time_string = "2025-01-14 14:30:00"
parsed_time = time.strptime(time_string, "%Y-%m-%d %H:%M:%S")
print("Parsed Time:", parsed_time)
```

---

#### **8. `time.mktime(t)`**
- Converts a `struct_time` object to epoch time.

**Usage:**
```python
import time

local_time = time.localtime()
epoch_time = time.mktime(local_time)
print("Epoch Time:", epoch_time)
```

## `io` module

The **`io`** module in Python provides tools for handling various types of I/O (Input/Output) streams. It is a fundamental module used to work with files, buffers, and streams in Python, offering a consistent interface for managing text and binary data.

---

### Key Features of the `io` Module

1. **File I/O**: Work with files in text or binary modes.
2. **Memory Streams**: Use in-memory streams like `StringIO` and `BytesIO`.
3. **Buffering**: Control buffering of input and output.
4. **Base Classes**: Define your custom I/O classes by extending the provided base classes.

---

### Key Classes in `io`

#### **1. `io.TextIOWrapper`**
- Used for handling text files.
- Handles encoding, decoding, and line-ending translation.

**Example: Reading and Writing Text Files**
```python
import io

# Writing to a text file
with io.open("example.txt", mode="w", encoding="utf-8") as file:
    file.write("Hello, world!")

# Reading from a text file
with io.open("example.txt", mode="r", encoding="utf-8") as file:
    content = file.read()
print(content)  # Hello, world!
```

---

#### **2. `io.BufferedReader`, `io.BufferedWriter`, and `io.BufferedRandom`**
- Handle binary files with buffering.
- `BufferedReader`: For reading binary data.
- `BufferedWriter`: For writing binary data.
- `BufferedRandom`: For both reading and writing binary data.

**Example: Reading and Writing Binary Files**
```python
import io

# Writing binary data
with io.open("example.bin", mode="wb") as file:
    file.write(b"\x00\x01\x02\x03")

# Reading binary data
with io.open("example.bin", mode="rb") as file:
    binary_data = file.read()
print(binary_data)  # b'\x00\x01\x02\x03'
```

---

#### **3. `io.StringIO`**
- Provides an in-memory text stream.
- Useful for manipulating text without involving actual files.

**Example: Using `StringIO`**
```python
from io import StringIO

# Creating a text stream
text_stream = StringIO("Initial text")

# Writing to the stream
text_stream.write("\nAdded text")

# Reading from the stream
text_stream.seek(0)  # Move to the beginning
print(text_stream.read())
# Output:
# Initial text
# Added text
```

---

#### **4. `io.BytesIO`**
- Provides an in-memory binary stream.
- Useful for manipulating binary data without actual files.

**Example: Using `BytesIO`**
```python
from io import BytesIO

# Creating a binary stream
binary_stream = BytesIO(b"Initial bytes")

# Writing to the stream
binary_stream.write(b"\nMore bytes")

# Reading from the stream
binary_stream.seek(0)  # Move to the beginning
print(binary_stream.read())
# Output: b'Initial bytes\nMore bytes'
```

---

### Common Modes in `io.open`

| **Mode** | **Description**                                                 |
|----------|-----------------------------------------------------------------|
| `r`      | Open for reading (default).                                    |
| `w`      | Open for writing; truncates the file if it exists.             |
| `x`      | Open for exclusive creation; fails if the file exists.         |
| `a`      | Open for writing; appends to the end of the file if it exists. |
| `b`      | Binary mode (e.g., `rb`, `wb`).                                |
| `t`      | Text mode (default, e.g., `rt`, `wt`).                         |
| `+`      | Open for updating (reading and writing).                       |

### Example: Reading and Writing Files

#### **Reading a Text File Line-by-Line**
```python
with io.open("example.txt", mode="r", encoding="utf-8") as file:
    for line in file:
        print(line.strip())  # Strip trailing newline characters
```

#### **Appending to a Text File**
```python
with io.open("example.txt", mode="a", encoding="utf-8") as file:
    file.write("\nAdditional text")
```

---

### Buffering in I/O Operations

Buffering can improve performance by reducing the number of I/O operations. The `io` module provides buffered I/O classes:

- **`BufferedReader`**: For reading binary data with buffering.
- **`BufferedWriter`**: For writing binary data with buffering.
- **`BufferedRandom`**: For reading and writing binary data.

**Example: Buffered Binary Read/Write**
```python
with io.BufferedWriter(io.open("example.bin", mode="wb")) as writer:
    writer.write(b"Buffered binary data")

with io.BufferedReader(io.open("example.bin", mode="rb")) as reader:
    data = reader.read()
print(data)  # b'Buffered binary data'
```

---

### Text Encoding and Decoding

#### Specifying Encoding
The `encoding` parameter in `io.open` allows you to handle various text encodings, such as UTF-8 or ISO-8859-1.

**Example: Writing with Encoding**
```python
with io.open("example.txt", mode="w", encoding="utf-16") as file:
    file.write("Hello, world!")
```

**Example: Reading with Encoding**
```python
with io.open("example.txt", mode="r", encoding="utf-16") as file:
    content = file.read()
print(content)  # Hello, world!
```

---

### Custom I/O Classes

You can create custom I/O classes by extending the base classes provided by `io`.

**Example: Custom Stream**
```python
from io import TextIOBase

class UpperCaseStream(TextIOBase):
    def __init__(self, stream):
        self.stream = stream

    def write(self, s):
        self.stream.write(s.upper())

# Using the custom stream
import sys
stream = UpperCaseStream(sys.stdout)
stream.write("hello, world!\n")  # Outputs: HELLO, WORLD!
```

### Use Cases of `io`

1. **File Handling**: Reading and writing text and binary files.
2. **In-Memory Streams**: Simulate file operations without disk access.
3. **Buffered I/O**: Efficient handling of large files.
4. **Custom I/O**: Create streams with custom behavior (e.g., compression, encryption).
5. **Text Encoding**: Handle files with different character encodings.

The `io` module is a fundamental tool for Python developers, offering both simplicity and flexibility for file and stream-based I/O operations.

## `zipfile` module

The **`zipfile`** module in Python provides tools to create, read, write, extract, and manage ZIP files. It supports both compression and decompression, making it a powerful tool for handling ZIP archives.

---

### Key Features of the `zipfile` Module

1. **Create ZIP Archives**: Compress files into a ZIP archive.
2. **Extract ZIP Archives**: Decompress and extract files from a ZIP archive.
3. **Read ZIP Contents**: Inspect the files in a ZIP archive.
4. **Add Compression**: Use `ZIP_DEFLATED` for compressed archives.

---

### Common Classes and Methods

#### **1. `zipfile.ZipFile`**
- The main class for working with ZIP files.

**Creating a ZIP File**
```python
import zipfile

with zipfile.ZipFile("example.zip", "w") as zf:
    zf.write("file1.txt")  # Add file1.txt to the ZIP archive
    zf.write("file2.txt")  # Add file2.txt to the ZIP archive
```

**Reading a ZIP File**
```python
with zipfile.ZipFile("example.zip", "r") as zf:
    print(zf.namelist())  # List of files in the archive
```

**Extracting All Files**
```python
with zipfile.ZipFile("example.zip", "r") as zf:
    zf.extractall("output_folder")  # Extract files to output_folder
```

---

#### **2. `zipfile.is_zipfile(filename)`**
- Checks if a file is a valid ZIP archive.

**Usage:**
```python
import zipfile

print(zipfile.is_zipfile("example.zip"))  # True if valid ZIP
```

---

#### **3. Modes in `ZipFile`**

| **Mode** | **Description**                               |
|----------|-----------------------------------------------|
| `'r'`    | Read-only mode (default).                    |
| `'w'`    | Write mode (creates a new ZIP file).         |
| `'x'`    | Exclusive creation (fails if the file exists).|
| `'a'`    | Append mode (adds files to an existing ZIP). |

---

### Compression Options

#### Compression Types

| **Constant**       | **Description**                                      |
|--------------------|------------------------------------------------------|
| `ZIP_STORED`       | No compression (default).                           |
| `ZIP_DEFLATED`     | Standard compression using the zlib library.        |
| `ZIP_BZIP2`        | Compression using the bzip2 library (Python 3.3+).  |
| `ZIP_LZMA`         | Compression using the LZMA library (Python 3.3+).   |

**Example: Using Compression**
```python
import zipfile

with zipfile.ZipFile("compressed.zip", "w", compression=zipfile.ZIP_DEFLATED) as zf:
    zf.write("file1.txt")
    zf.write("file2.txt")
```

---

### Example Workflows

#### **1. Creating a ZIP File**
```python
import zipfile

files = ["file1.txt", "file2.txt", "file3.txt"]

with zipfile.ZipFile("archive.zip", "w") as zf:
    for file in files:
        zf.write(file)
print("ZIP archive created!")
```

---

#### **2. Extracting Files**
```python
import zipfile

with zipfile.ZipFile("archive.zip", "r") as zf:
    zf.extractall("extracted_files")
print("Files extracted!")
```

---

#### **3. Adding Files to an Existing ZIP**
```python
import zipfile

with zipfile.ZipFile("archive.zip", "a") as zf:
    zf.write("new_file.txt")
print("File added to archive!")
```

---

#### **4. Reading File Info**
```python
import zipfile

with zipfile.ZipFile("archive.zip", "r") as zf:
    for file_info in zf.infolist():
        print(f"Name: {file_info.filename}")
        print(f"Size: {file_info.file_size} bytes")
        print(f"Compressed Size: {file_info.compress_size} bytes")
```

---

### Example: Compressing a Folder

#### Compressing an Entire Folder
```python
import os
import zipfile

def zip_folder(folder_path, zip_name):
    with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                zf.write(file_path, os.path.relpath(file_path, folder_path))

zip_folder("my_folder", "my_folder.zip")
print("Folder compressed!")
```

---

### Example: Extracting Specific Files

#### Extract Only Certain Files
```python
import zipfile

with zipfile.ZipFile("archive.zip", "r") as zf:
    files_to_extract = ["file1.txt", "file2.txt"]
    for file in files_to_extract:
        zf.extract(file, "specific_files")
print("Specific files extracted!")
```

---

### Notes and Limitations

1. **Supported File Formats**: `zipfile` only works with ZIP files. Use libraries like `tarfile` for TAR archives.
2. **Password Protection**: While `zipfile` supports password protection (`setpassword`), it only works with legacy encryption methods.
3. **File Size**: Handles large files, but consider using streaming for very large archives.

---

### Summary of Key Methods

| **Method**           | **Description**                                         |
|-----------------------|---------------------------------------------------------|
| `ZipFile.write()`     | Adds a file to the archive.                            |
| `ZipFile.namelist()`  | Returns a list of filenames in the archive.            |
| `ZipFile.extractall()`| Extracts all files from the archive.                   |
| `ZipFile.extract()`   | Extracts a specific file from the archive.             |
| `ZipFile.infolist()`  | Returns information about files in the archive.        |
| `ZipFile.close()`     | Closes the archive (automatically handled with `with`).|

---

### Use Cases for the `zipfile` Module

1. **Backup Files**: Compress files into ZIP archives for backups.
2. **Distribute Data**: Package multiple files into a single ZIP for easy sharing.
3. **Read ZIP Archives**: Inspect and extract files from downloaded ZIPs.
4. **Log Compression**: Archive old logs to save space.

The `zipfile` module is versatile and an essential tool for file compression and management in Python.

## `tarfile` module

The **`tarfile`** module in Python provides tools to work with TAR (Tape Archive) files. It supports creating, reading, extracting, and modifying TAR archives, with optional compression (e.g., Gzip, Bzip2).

---

### Key Features of the `tarfile` Module

1. **Create TAR Archives**: Combine multiple files and directories into a single TAR file.
2. **Extract TAR Archives**: Extract files and directories from a TAR file.
3. **Compressed Archives**: Work with compressed TAR files (e.g., `.tar.gz`, `.tar.bz2`).
4. **Read/Write Modes**: Read and write archives in various modes (text or binary).
5. **Selective Extraction**: Extract specific files from TAR archives.

---

### Commonly Used Functions and Classes

#### **1. `tarfile.open(name, mode)`**
- Opens a TAR file for reading, writing, or appending.
- **`name`**: Path to the TAR file.
- **`mode`**: Specifies the operation mode.

**Modes in `tarfile.open`:**

| **Mode** | **Description**                                         |
|----------|---------------------------------------------------------|
| `'r'`    | Read-only mode for uncompressed TAR files.              |
| `'r:gz'` | Read-only mode for Gzip-compressed TAR files (`.tar.gz`).|
| `'r:bz2'`| Read-only mode for Bzip2-compressed TAR files (`.tar.bz2`).|
| `'w'`    | Write mode for creating uncompressed TAR files.         |
| `'w:gz'` | Write mode for creating Gzip-compressed TAR files.      |
| `'w:bz2'`| Write mode for creating Bzip2-compressed TAR files.     |
| `'x'`    | Extract-only mode for uncompressed TAR files.           |

---

### Creating TAR Archives

#### **Create an Uncompressed TAR Archive**
```python
import tarfile

with tarfile.open("example.tar", "w") as tar:
    tar.add("file1.txt")  # Add individual files
    tar.add("folder")     # Add an entire folder
print("TAR archive created!")
```

#### **Create a Gzip-Compressed TAR Archive**
```python
with tarfile.open("example.tar.gz", "w:gz") as tar:
    tar.add("file1.txt")
print("Compressed TAR archive created!")
```

---

### Extracting TAR Archives

```python
import tarfile
```

#### **Extract Specific Files**
```python
with tarfile.open("example.tar", "r") as tar:

```

### Example: Compressing and Extracting a Folder

#### **Compressing a Folder**
```python
import tarfile

def compress_folder(folder_name, tar_name):
    with tarfile.open(tar_name, "w:gz") as tar:
        tar.add(folder_name, arcname=".")
    print(f"Folder '{folder_name}' compressed into '{tar_name}'")

compress_folder("my_folder", "my_folder.tar.gz")
```

#### **Extracting a Folder**
```python
import tarfile

def extract_folder(tar_name, output_folder):
    with tarfile.open(tar_name, "r:gz") as tar:
        tar.extractall(output_folder)
    print(f"Archive '{tar_name}' extracted to '{output_folder}'")

extract_folder("my_folder.tar.gz", "extracted_folder")
```

### Selective Extraction

**Extract Files Matching a Pattern**
```python
import tarfile

with tarfile.open("example.tar", "r") as tar:

        if member.name.endswith(".txt"):
            tar.extract(member, "output_folder")
```

### Notes and Limitations

   - Supports `gzip` (`.tar.gz`) and `bzip2` (`.tar.bz2`) natively.

2. **Performance**:
   - Use compressed formats only when needed, as they are slower than uncompressed archives.
3. **Selective Operations**:
   - TAR archives do not support random access; extracting individual files can be slow for large archives.
